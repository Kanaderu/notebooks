This notebook looks at approximating the impulse response of a synapse digitally and asynchronously. 

Specifically:

-What value should the synapse increment by when receiving a spike?

-How should the synapse handle spikes that arrive between synapse ticks.

```python
def h_exp(time, t_offset, tau):
    """Defines the impulse response of a synapse"""
    N = len(time)
    h = zeros(N)
    idx = time >= t_offset
    h[idx] = 1./tau*np.exp(-(time[idx]-t_offset)/tau)
    return h
```

```python
%pylab inline
# simulation parameters
tau = .1  # filter time constant
dt = tau/5.  # time step of the filter
dtsim = dt/100.  # time step of this simulation
nperiods = 3  # number of time constants to simulate
```

The impulse response of our low pass filter is a decaying exponential.

```python
ncurves = 5
offsets = linspace(0, dt, ncurves)
time = arange(0, dt*nperiods, dtsim)
figure('offsets', figsize=(8,6))
for ti in offsets:
    plot(time, h_exp(time, ti, tau))
periods = arange(nperiods+1)*dt
ylims = [-1, 1./tau+1]
ylim(ylims)
vlines(periods, ylims[0], ylims[1], linestyle='dotted')
hlines(0, xlim()[0], xlim()[1]) 
title('Here are some shifted impulse responses', fontsize=16)
xlabel('time(s)', fontsize=16)
```

# Setting the synapse output

We can use sample the continuous waveform to a produce a discrete waveform. This is the impulse invariance technique.

```python
impulse = h_exp(time, 0, tau)
h_samp = h_exp(periods, 0 , tau)

figure('impulse inv sample', figsize=(8,6))
plot(time, impulse, label='impulse response')
plot(periods, h_samp, 'go', label='samples')
ylim(ylims)
vlines(periods, ylims[0], ylims[1], linestyle='dotted')
hlines(0, xlim()[0], xlim()[1]) 
legend(loc='center left')
title('Here is what we would sample from an impulse', fontsize=16)
xlabel('time(s)', fontsize=16)
```

The impulse invariance technique causes problems because we'll be holding the synapse output at a sample value until the next sample. Since the synapse would decay over the period but our output does not, holding the synapse at the sample value consistently overestimates the synapse value. 

```python
output = np.zeros(time.shape)
for p_idx in xrange(len(periods[:-1])):
    idx = np.logical_and(time>=periods[p_idx], time<periods[p_idx+1])
    output[idx] = h_samp[p_idx]

figure('impulse inv hold', figsize=(8,6))
plot(time, impulse, label='impulse response')
plot(periods, h_samp, 'go', label='samples')
plot(time, output, 'r', label='output')
ylim(ylims)
vlines(periods, ylims[0], ylims[1], linestyle='dotted')
hlines(0, xlim()[0], xlim()[1]) 
legend(loc='center left')
title('If we hold the output at the sample value,\nthen output will be more powerful than impulse response', fontsize=16)
xlabel('time(s)', fontsize=16)
```

We can compensate for the output overshoot by using the mean value of the impulse response the period.

The impulse response of a synaptic filter with time constant $\tau$ and sample period $\Delta t$ is 
$$h(t) = \frac{1}{\tau}e^{-t/\tau}u(t)$$

where $u(t)$ is the unit step response.

The mean value, $m$, of the first sample period is

```python
%%latex
\begin{aligned}
m &= \frac{1}{\Delta t}\int_0^{\Delta t}h(t)dt \\
  &= \frac{1}{\Delta t}\int_0^{\Delta t}\frac{1}{\tau}e^{-t/\tau}dt \\
  &= \frac{1}{\Delta t}\frac{1}{\tau}\left[-\tau e^{-t/\tau}\right]_0^{\Delta t} \\
  &= \frac{1}{\Delta t}\left[e^{-t/\tau}\right]_{\Delta t}^0 \\
m &= \frac{1}{\Delta t}\left(1-e^{-\Delta t / \tau}\right)
\end{aligned}
```

In addition, the mean value, $m_i$, for any sample, $i$, can be expressed as $e^{-i\Delta t/\tau}m$

```python
%%latex
\begin{aligned}
m_i &= \frac{1}{\Delta t}\int_{i\Delta t}^{(i+1)\Delta t}h(t)dt \\
    &= \frac{1}{\Delta t}\int_{i\Delta t}^{(i+1)\Delta t}\frac{1}{\tau}e^{-t/\tau}dt \\
    &= \frac{1}{\Delta t}\left[e^{-t/\tau}\right]_{(i+1)\Delta t}^{i\Delta t} \\
    &= \frac{1}{\Delta t}\left(e^{-i\Delta t / \tau}-e^{-(i+1)\Delta t / \tau}\right) \\
    &= \frac{1}{\Delta t}\left(1-e^{-\Delta t / \tau}\right)e^{-i\Delta t / \tau} \\
m_i &= e^{-i\Delta t / \tau}m
\end{aligned}
```

Using the mean value instead of the impulse response instead of the sample points as the output preserves the overall area of the impulse response. The original impulse response has an area of 1:

```python
%%latex
\begin{aligned}
\int_0^\infty \frac{1}{\tau}e^{-t/\tau}dt &= \left[e^{-t/\tau}\right]_{\infty}^0 \\
&= 1-0 \\
&= 1
\end{aligned}
```

The area of the mean value approximation is also $1$:

```python
%%latex
\begin{aligned}
\sum_{i=0}^\infty m_i \Delta t &= \sum_{i=0}^\infty e^{-i\Delta t / \tau}m \Delta t  \\
&= m\Delta t\sum_{i=0}^\infty e^{-i\Delta t / \tau} \\
&= m\Delta t \frac{1}{1-e^{-\Delta t/\tau}} \\
&= \frac{1}{\Delta t}\left(1-e^{-\Delta t / \tau}\right) \Delta t \frac{1}{1-e^{-\Delta t/\tau}} \\
&= 1
\end{aligned}
```

```python
def h_mv0(dt, tau):
    """Computes the mean value of the impulse during the 
    initial sampling period"""
    mv0 = 1./dt*(1.-exp(-dt/tau))
    return mv0

m0 = h_mv0(dt, tau)
h_mv = m0
h_mv = h_mv*exp(-dt/tau)**arange(nperiods+1)

output = np.zeros(time.shape)
for p_idx in xrange(len(periods[:-1])):
    idx = np.logical_and(time>=periods[p_idx], time<periods[p_idx+1])
    output[idx] = h_mv[p_idx]

figure('mean value hold', figsize=(8,6))
plot(time, impulse, label='impulse response')
plot(time, output, 'r', label='output')
ylim(ylims)
vlines(periods, ylims[0], ylims[1], linestyle='dotted')
hlines(0, xlim()[0], xlim()[1]) 
legend(loc='center left')
title('Output as the mean value', fontsize=16)
xlabel('time(s)', fontsize=16)
```

```python
impulse_int = 1-exp(-time/tau)
output_mv_int = zeros(periods.shape)

output_mv_int[1:] = cumsum(h_mv[:-1])*dt

figure('area preservation', figsize=(8,6))
plot(time, impulse_int, label='impulse response')
plot(periods, output_mv_int, 'g', label='output')
ylim([-.01,.5])
vlines(periods, ylim()[0], ylim()[1], linestyle='dotted')
hlines(0, xlim()[0], xlim()[1]) 
legend(loc='upper left')
title('Setting the output to the mean value preserves the area', fontsize=16)
ylabel('cumulative area', fontsize=16)
xlabel('time(s)', fontsize=16)
```

# Handling Asynchronous inputs

The second problem we deal with is that spikes will arrive asynchronously, regardless of the $\Delta t$ of the synapse. At every $\Delta t$, the synapse will scale it's value by $e^{-\Delta t/\tau}$. Spikes that arrive between samples have not had their initial output applied for $\Delta t$ and so are prematurely scaled.

```python
offset = offsets[2]
impulse = h_exp(time, offset, tau)

output = np.zeros(time.shape)
for p_idx in xrange(len(periods[:-1])):
    idx = np.logical_and(time>=periods[p_idx], time<periods[p_idx+1])
    idx[time<offset] = False
    output[idx] = h_mv[p_idx]

figure('area loss shifting', figsize=(8,6))
plot(time, impulse, label='impulse response')
plot(time, output, 'r', label='output')
ylim(ylims)
vlines(periods, ylims[0], ylims[1], linestyle='dotted')
hlines(0, xlim()[0], xlim()[1]) 
legend(loc='lower right')
title('Output for spikes that arrive between samples\nundershoot the impulse', fontsize=16)
xlabel('time(s)', fontsize=16)
```

 To handle this, we scale up the output value to compensate for the lost area. Instead of setting the initial output to the mean value $m$, set the output to $mc$, where $c$ scales up the output to compensate for the lost area. For time offset $t_i$, we solve for $c$ to preserve the area (of 1).

```python
%%latex
\begin{aligned}
1 &= mc(\Delta t - t_i) + \sum_{i=1}^\infty mc e^{-i\Delta t/\tau}\Delta t \\
  &= mc(\Delta t - t_i) + mc\frac{e^{-\Delta t/\tau}}{1-e^{-\Delta t/\tau}}\Delta t \\
  &= mc(\Delta t - t_i) + c\frac{1}{\Delta t}\left(1-e^{-\Delta t / \tau}\right)\frac{e^{-\Delta t/\tau}}{1-e^{-\Delta t/\tau}}\Delta t \\
  &= mc(\Delta t - t_i) + ce^{-\Delta t/\tau} \\
  &= c\left(m(\Delta t - t_i) + e^{-\Delta t/\tau}\right) \\
c &=\left(m(\Delta t - t_i) + e^{-\Delta t/\tau}\right)^{-1}
\end{aligned}
```

```python
def async_comp(ti, tau, dt):
    c = (h_mv0(dt, tau) * (dt - ti) + exp(-dt / tau))**-1
    return c

ti = linspace(0, dt, 10)
c = async_comp(ti, tau, dt)
figure('async compensation factor', figsize=(8,6))
plot(ti/dt, c)
ylabel(r'$c$', fontsize=18)
xlabel(r'$t_i / \Delta t$', fontsize=18)
title(r'compensation factor as time of spike within $\Delta t$', fontsize=16)
```

```python
c = async_comp(tau=tau, dt=dt, ti=offset)
print c
h_mv = m0*c
h_mv = h_mv*exp(-dt/tau)**arange(nperiods+1)

output_corrected = np.zeros(time.shape)
for p_idx in xrange(len(periods[:-1])):
    idx = np.logical_and(time>=periods[p_idx], time<periods[p_idx+1])
    idx[time<offset] = False
    output_corrected[idx] = h_mv[p_idx]
    
figure('area restoriation', figsize=(8,6))
plot(time, impulse, label='impulse response')
plot(time, output, 'r:', label='output uncorrected')
plot(time, output_corrected, 'r', label='output corrected')
ylim(ylims)
vlines(periods, ylims[0], ylims[1], linestyle='dotted')
hlines(0, xlim()[0], xlim()[1]) 
legend(loc='lower right')
title('Scaling the output to compensate for area lost\n of a spike arriving between samples', fontsize=16)
xlabel('time(s)', fontsize=16)
```

```python
from scipy.integrate import cumtrapz

h_area = cumtrapz(impulse, time, initial=0)
out_area = cumtrapz(output, time, initial=0)
out_corr_area = cumtrapz(output_corrected, time, initial=0)
figure('corrected areas', figsize=(8,6))
plot(time, h_area, label='impulse response')
plot(time, out_area, 'r:', label='output uncorrected')
plot(time, out_corr_area, 'r', label='output corrected')
ylim(-.01, .45)
vlines(periods, ylim()[0], ylim()[1], linestyle='dotted')
hlines(0, xlim()[0], xlim()[1]) 
legend(loc='upper left')
title('Scaled output matches area of impulse', fontsize=16)
ylabel('cumulative area', fontsize=16)
xlabel('time(s)', fontsize=16)
```

However, in hardware, it would be too expensive to calculate the compensation factor for each spike. To overcome this, if we assume that spikes are Poisson, then we assume that spikes arrive uniformly within the sampling period. With this assumption, we can calculate the expected compensation factor to apply to all spikes so that, on average, each spike response has the same area of the impulse response.

```python
%%latex
\begin{aligned}
c &=\left(m(\Delta t - t_i) + e^{-\Delta t/\tau}\right)^{-1} \\
E[c] &= E\left[\left(m(\Delta t - t_i) + e^{-\Delta t/\tau}\right)^{-1}\right] \\
 &= \frac{1}{\Delta t}\int_0^{\Delta t} \left(m(\Delta t - t_i) + e^{-\Delta t/\tau}\right)^{-1} dt_i
\end{aligned}
```

```python
from scipy.integrate import quad

ti = linspace(0, dt, 1000)
mean_c = cumtrapz(async_comp(ti, tau=tau, dt=dt), ti, initial=0)[-1]/dt
print 'mean correction scaling factor with cumtrapz:', mean_c

int_res, abserr = quad(async_comp, 0, dt, (tau, dt))
mean_c = int_res / dt
print 'mean correction scaling factor with quad:', mean_c
```
