# Notebook description

This notebook compares controlling the dynamics of a network using recurrent connections at the NEF level versus using adaptation at the single neuron level

# Summary of results

 - Both the NEF dynamics formulation and the adaptive neuron can increase the speed of a network
 - Whereas the NEF dynamics formulation does not change the spike rate, adaptation can decrease the spike rate.
 - We still don't have a clean mapping between the parameters of adaptation and their effects on the dynamics of a network

# Simulation setup

```python
%pylab inline
import numpy as np
import nengo
from nengo.utils.functions import piecewise
from nengo.utils.numpy import rms, rmse
from nengo.synapses import filtfilt

def square_wave(t, period=.1, amplitude=.5):
    return amplitude*(2.*(int(t/period) % 2) - 1.)
```

# Feedfoward dynamics are governed by the synaptic time constant

For the feedfoward NEF network

![feedfoward network](figures/feed_foward_nengo.png "feedfoward network")

the dynamics of $x(t)$ are governed by the synaptic filter $h(t)$. If we decode identity from the Ensemble, then this network implements 

\begin{align}
x(t) &= h(t)*u(t) \\
X(s) &= H(s)U(s) \\
\end{align}

Note that the Node is connected directly to the Ensemble without a synapse. 

```python
def build_ff_sys(nneurons, dim, pstc_out, radius=1., pstc_in=None):
    net = nengo.Network(seed=1)
    with net:
        net.ens = nengo.Ensemble(nneurons, dim, radius=radius, seed=1)
        net.input = nengo.Node(None, size_in=dim)
        net.output = nengo.Node(None, size_in=dim)
        nengo.Connection(net.input, net.ens, synapse=pstc_in)
        nengo.Connection(net.ens, net.output, synapse=pstc_out)
        net.probe_spikes = nengo.Probe(net.ens.neurons)
    return net
```

```python
def test_ff_dynamics_1syn(pstc):
    dt = .001
    net = build_ff_sys(nneurons=100, dim=1, pstc_out=pstc)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=None)

        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_out = nengo.Probe(net.output, synapse=None)
        probe_pstc = nengo.Probe(stimulus, synapse=pstc)

    sim = nengo.Simulator(net, dt=dt)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_pstc])
    sim_rms = rms(sim.data[probe_pstc])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_out], 'b', label='neural ' + r'$x(t)$')
    plot(t, sim.data[probe_pstc], 'r', label='predicted ' + r'$x(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (np.mean(spike_dat)))
    fig.subplots_adjust(top=.85)
    suptitle(r'$\tau_{syn}=%.3f$' % pstc, fontsize=14)

pstcs = [.001, .01, .1]
for pstc in pstcs:
    test_ff_dynamics_1syn(pstc=pstc)
```

With the random seed fixed of the Network and Ensemble fixed, the Ensemble's spike times are identical as expected. Therefore, the dynamics of $x(t)$ are governed solely by the output synaptic filter.

If you include a synapse in Connection between the input Node and the Ensemble, 

![feedfoward network with input synapse](figures/feed_foward_nengo_2syn.png "feedfoward network with input synapse")

the output will be a double filtered version of the input.

\begin{align}
x(t) &= h(t) * h(t) * u(t) \\
X(s) &= H(s)H(s)U(s) \\
\end{align}

Note that this is not the same filtering with twice the synaptic time constant. For $H(s)=\frac{1}{\tau s+1}$

\begin{align}
X(s) &= \frac{1}{\tau s+1}\frac{1}{\tau s+1}U(s) \\
 &= \frac{1}{\tau^2s^2+2\tau s+1}U(s) \\
 &\neq \frac{1}{2\tau s+1}U(s) \\
\end{align}

We see that $X(s)$ is now described by a second order system.

```python
def test_ff_dynamics_2syn(pstc):
    dt = .001
    net = build_ff_sys(nneurons=100, dim=1, pstc_out=pstc)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=pstc)
        node1 = nengo.Node(lambda t, x:x, size_in=1)
        nengo.Connection(stimulus, node1, synapse=pstc)
        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_out = nengo.Probe(net.output, synapse=None)
        probe_pstc = nengo.Probe(node1, synapse=pstc)
    sim = nengo.Simulator(net, dt=dt)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_pstc])
    sim_rms = rms(sim.data[probe_pstc])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_out], 'b', label='neural ' + r'$x(t)$')
    plot(t, sim.data[probe_pstc], 'r', label='predicted ' + r'$x(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (np.mean(spike_dat)))
    fig.subplots_adjust(top=.85)
    suptitle(r'$\tau_{syn}=%.3f$' % pstc, fontsize=14)

pstcs = [.001, .01, .1]
for pstc in pstcs:
    test_ff_dynamics_2syn(pstc=pstc)
```

We see that the predicted second order dynamics matches the observed neural $x(t)$ dynamics. Unlike the network with only an output synapse, the Ensemble in this network spikes at differing rates.

# We can use NEF to control the system dynamics

NEF uses recurrence to create the neural system

![neurally plausible dynamical system](figures/neural_system.png "neurally plausible dynamical system")

where

\begin{align}
x(t) &= h(t)*[A'x(t)+B'u(t)] \\
X(s) &= H(s)[A'X(s)+B'U(s)] \\
\end{align}

To implement dynamics specified by

\begin{align}
\dot{x}(t) &= Ax(t)+Bu(t) \\
sX(s) &= AX(s)+BU(s) \\
\end{align}

we use

\begin{align}
A' &= \tau A+I \\
B' &= \tau B \\
\end{align}

To implement the neural system with nengo, we use

![NEF dynamical network](figures/dyn_nengo.png "NEF dynamical network")

where the decoders are trained for identity. The input to the Ensemble as well as the instaneous value of the decoded spikes both represent $Ax(t)+Bu(t)$. For the double filtering reason described in the feedfoward cases, we do not use a synapse in the Connection between the Node and the Ensemble.

To implement a system with dynamics

\begin{align}
\tau_{sys}\dot{x}(t) &= -x(t)+u(t) \\
\end{align}

we use

\begin{align}
A' &= -\frac{\tau}{\tau_{sys}}+1 \\
B' &= \frac{\tau}{\tau_{sys}} \\
\end{align}

```python
def build_dyn_sys(nneurons, dim, pstc, A, B, radius=1., pstc_in=None):
    A = pstc*A+np.eye(A.shape[0])
    B = pstc*B
    net = nengo.Network(seed=1)
    with net:
        net.ens = nengo.Ensemble(nneurons, dim, radius=radius, seed=1)
        net.input = nengo.Node(None, size_in=dim)
        net.output = nengo.Node(None, size_in=dim)
        nengo.Connection(net.input, net.ens, transform=B, synapse=pstc_in)
        nengo.Connection(net.ens, net.ens, transform=A, synapse=pstc)
        nengo.Connection(net.ens, net.output, synapse=pstc)
        net.probe_spikes = nengo.Probe(net.ens.neurons)
    return net, A, B
```

```python
def test_r_dynamics(pstc, tau_sys, radius):
    dt = .001
    A = np.array([[-1./tau_sys]])
    B = np.array([[1./tau_sys]])
    net, Ap, Bp = build_dyn_sys(nneurons=100, dim=1, pstc=pstc, A=A, B=B, radius=radius)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=None)
        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_ens_in = nengo.Probe(net.ens, 'input')
        probe_out = nengo.Probe(net.output, synapse=None)
        probe_tau_sys = nengo.Probe(stimulus, synapse=tau_sys)
    sim = nengo.Simulator(net)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_tau_sys])
    sim_rms = rms(sim.data[probe_tau_sys])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_ens_in], c='k', alpha=.4, label='ens. input')
    plot(t, sim.data[probe_out], 'b', label='neural ' + r'$x(t)$')
    plot(t, sim.data[probe_tau_sys], 'r', label='predicted ' + r'$x(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (np.mean(spike_dat)))
    fig.subplots_adjust(top=.8)
    suptitle(r"$\tau_{syn}=%.3f$ $\tau_{sys}=%.3f$ $A'=%.2f$ $B'=%.2f$" % (pstc, tau_sys, Ap[0], Bp[0]), fontsize=16)
    
tau_sys = .01
pstcs = [.005, .01, .02]
radii = [1., 1., 1.5]
for pstc, radius in zip(pstcs, radii):
    test_r_dynamics(pstc, tau_sys, radius)
```

The network implements the predicted dynamics. However, the firing rate does not decrease below 8000 spks/s with faster system dynamics. Also note how spiky the Ensemble input becomes as the system time constant drops below the synaptic time constant. To accomodate the excursions of the Ensemble input, we must increase the Ensemble radius.

The Ensemble's input represents $A'x(t)+B'u(t)$, so the large spikes in the Ensemble's input result from a large input gain $B'=\frac{\tau}{\tau_{sys}}$. As $\tau_{sys}$ decreases relative to the synaptic $\tau$, $B'$ increases. When $B'$ is large, a sudden change in system input $u(t)$ produces a large change in the Ensemble's input until the decoded $x(t)$ can catch up.

## Avoid large input transients by rescaling the system

Imagine that instead of implementing the low-pass filter

\begin{align}
\tau_{sys}\dot{x} &= -x+u \\
\dot{x} &= -\frac{1}{\tau_{sys}}x+\frac{1}{\tau_{sys}}u \\
\end{align}

we implemented 

\begin{align}
\tau_{sys}\dot{x} &= -x+u \\
\dot{x} &= -\frac{1}{\tau_{sys}}x+u \\
\end{align}

and took the output of the system to be

\begin{align}
y &= Cx \\
\end{align}

where $C=\frac{1}{\tau_{sys}}$ then

\begin{align}
\dot{y} &= C\dot{x} \\
 &= \frac{1}{\tau_{sys}}\left(-\frac{1}{\tau_{sys}}x+u\right) \\
 &= \frac{1}{\tau_{sys}}\left(-y+u\right) \\
 &= -\frac{1}{\tau_{sys}}y+\frac{1}{\tau_{sys}}u \\
\tau_{sys}\dot{y} &= -y+u \\
\end{align}

which is exactly the dynamical equation of the original low-pass filter. To implement this system using neurons, we use 

\begin{align}
A' &= -\frac{\tau}{\tau_{sys}}+1 \\
B' &= \tau \\
C &= \frac{1}{\tau_{sys}} \\
\end{align}

and build

![reshuffled NEF dynamical system](figures/dyn_nengo_reshuffled.png "reshuffled NEF dynamical system")

```python
def test_r_dynamics_reshuffled(pstc, tau_sys, radius):
    dt = .001
    A = np.array([[-1./tau_sys]])
    B = np.array([[1.]])
    C = np.array([[1./tau_sys]])
    net, Ap, Bp = build_dyn_sys(nneurons=100, dim=1, pstc=pstc, A=A, B=B, radius=radius)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=None)
        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_ens_in = nengo.Probe(net.ens, 'input')
        node_out = nengo.Node(lambda t, x: x, size_in=1)
        nengo.Connection(net.output, node_out, transform=C, synapse=None)
        probe_out = nengo.Probe(node_out, synapse=None)
        probe_tau_sys = nengo.Probe(stimulus, synapse=tau_sys)
    sim = nengo.Simulator(net)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_tau_sys])
    sim_rms = rms(sim.data[probe_tau_sys])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_ens_in], c='k', alpha=.4, label='ens. input')
    plot(t, sim.data[probe_out], 'b', label='neural ' + r'$y(t)$')
    plot(t, sim.data[probe_tau_sys], 'r', label='predicted ' + r'$y(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (np.mean(spike_dat)))
    fig.subplots_adjust(top=.8)
    suptitle(
        r"$\tau_{syn}=%.3f$ $\tau_{sys}=%.3f$ $A'=%.2f$ $B'=%.3f$ $C=%.2f$" % (pstc, tau_sys, Ap[0], Bp[0], C[0]),
        fontsize=16)
    
tau_sys = .01
pstcs = [.005, .01, .02]
radii = [.016, .016, .016]
for pstc, radius in zip(pstcs, radii):
    test_r_dynamics_reshuffled(pstc, tau_sys, radius)
```

In some sense, we've traded one problem for another. The Ensemble inputs are now tiny because $B'$ is small, so we reduce the radius of the Ensemble.

## Adding an input synapse slows the system

Returning to the issue of not having an input synapse, if we connected the input Node to the Ensemble with a synapse, we would have

![dynamical NEF network with input synapse](figures/dyn_nengo_in_syn.png "dynamical NEF network with input synapse")

If we picked $A'$ and $B'$ to implement a low-pass filter $h_{sys}(t)$ with time constant $\tau_{sys}$, adding the input synapse would change our observed dynamics to be

\begin{align}
x(t) &= h_{sys}(t)*h(t)*u(t) \\
X(s) &= H_{sys}(s)H(s)U(s) \\
\end{align}

# We can use adaptive neurons to speed up the dynamics

The adaptive neuron is being analyzed [here](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/neuron/adaptive_LIF_neuron.ipynb). 

```python
def build_aff_sys(nneurons, dim, pstc_out, radius=1., pstc_in=None):
    neuron_type = nengo.AdaptiveLIF(tau_f=pstc_out, alpha_f=1.)
    net = nengo.Network(seed=1)
    with net:
        net.ens = nengo.Ensemble(
            nneurons, dim, radius=radius, neuron_type=neuron_type, seed=1)
        net.input = nengo.Node(None, size_in=dim)
        net.output = nengo.Node(None, size_in=dim)
        nengo.Connection(net.input, net.ens, synapse=pstc_in)
        nengo.Connection(net.ens, net.output, synapse=pstc_out)
        net.probe_spikes = nengo.Probe(net.ens.neurons)
    return net
```

```python
def test_aff_dynamics(pstc):
    dt = .001
    net = build_aff_sys(nneurons=100, dim=1, pstc_out=pstc)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=None)

        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_out = nengo.Probe(net.output, synapse=None)
        probe_pstc = nengo.Probe(stimulus, synapse=pstc)

    sim = nengo.Simulator(net, dt=dt)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_pstc])
    sim_rms = rms(sim.data[probe_pstc])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_out], 'b', label='neural ' + r'$x(t)$')
    plot(t, sim.data[probe_pstc], 'r', label='predicted ' + r'$x(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (np.mean(spike_dat)))
    fig.subplots_adjust(top=.85)
    suptitle(r'$\tau=%.3f$' % pstc, fontsize=14)

pstcs = [.001, .01, .1]
for pstc in pstcs:
    test_aff_dynamics(pstc=pstc)
```

Needs some debugging, but the spike rates drop as expected.

# Open questions and discussion

While we can reliably predict the firing rates of the adaptive neurons, we still need to know how the adaptation parameters determine the dynamics of the system. Only then can we quickly construct a network implementing a user specified time constant.

The adaptive neuron is difficult to analyze, but it seems to show second order behavior. Why not design a second order synapse that mimics the dynamic behavior of the adaptive neuron while being analytically tractable?
