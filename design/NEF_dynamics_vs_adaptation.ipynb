# Notebook description

This notebook compares controlling the dynamics of a network using recurrent connections at the NEF level versus using [adaptation](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/neuron/adaptive_LIF_neuron.ipynb) at the individual neuron level. Specifically we would like to know:
 - Can we use NEF dynamics to reduce the required spike rate to simultaneously meet a specified SNR and system time constant?
 - Can we use adaptive neurons to do the same?
 - Which method is better?

# Summary of results

 - Using recurrence and NEF dynamics formulation only increases the required spike rate to meet a SNR/system time constant specification.
  - A feedfoward network with a synapse matching the desired system time constant should always be less noisy.
 - Adaptive neurons can help reduce the required spike rate.
 - There is still work to be done on mapping the adaptive feedback parameters to the resulting system time constant

# Simulation setup

```python
%pylab inline
import numpy as np
import nengo
from nengo.utils.functions import piecewise
from nengo.utils.numpy import rms, rmse
from nengo.synapses import filtfilt
from nengo.dists import Uniform

def square_wave(t, period=.1, amplitude=.5):
    return amplitude*(2.*(int(t/period) % 2) - 1.)
```

# Feedfoward dynamics are governed by the synaptic time constant

We'll start with the basics -- a feedfoward NEF network with no recurrent connections and using standard LIF neurons.

![feedfoward network](figures/feed_foward_nengo.png "feedfoward network")

Here we will show that the dynamics of $x(t)$ are governed by the synaptic filter $h(t)$. If we decode identity from the Ensemble, then this network implements 

\begin{align}
x(t) &= h(t)*u(t) \\
X(s) &= H(s)U(s) \\
\end{align}

Note that the Node is connected directly to the Ensemble without a synapse. If we used a syanpse in the Connection from the Node to the Ensemble, we would observe a double filtered version of the input (see [Appendix](#A-feedfoward-network-with-two-synapses-double-filters-the-input)). Here, we will vary the synaptic time constant and see that the observed dynamics are governed by the synaptic time constant.

```python
def build_ff_sys(nneurons, dim, pstc_out, radius=1., pstc_in=None, max_rates=Uniform(200,400)):
    """Build a feedfoward network (communication channel)"""
    net = nengo.Network(seed=1)
    with net:
        net.ens = nengo.Ensemble(nneurons, dim, radius=radius, max_rates=max_rates, seed=1)
        net.input = nengo.Node(None, size_in=dim)
        net.output = nengo.Node(None, size_in=dim)
        nengo.Connection(net.input, net.ens, synapse=pstc_in)
        nengo.Connection(net.ens, net.output, synapse=pstc_out)
        net.probe_spikes = nengo.Probe(net.ens.neurons)
    return net
```

```python
def test_ff_dynamics_1syn(nneurons, pstc, max_rates):
    dt = .001
    net = build_ff_sys(nneurons=nneurons, dim=1, pstc_out=pstc, max_rates=max_rates)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=None)

        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_out = nengo.Probe(net.output, synapse=None)
        probe_pstc = nengo.Probe(stimulus, synapse=pstc)

    sim = nengo.Simulator(net, dt=dt)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_pstc])
    sim_rms = rms(sim.data[probe_pstc])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_out], 'b', label='observed ' + r'$x(t)$')
    plot(t, sim.data[probe_pstc], 'r', label='predicted ' + r'$x(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    mean_rates = np.mean(spike_dat)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (mean_rates))
    fig.subplots_adjust(top=.85)
    suptitle(r'$\tau_{syn}=%.3f$' % pstc, fontsize=14)
    return mean_rates

nneuronss = [900, 200, 100, 100, 100, 100, 100]
pstcs = [.001, .002, .005, .01, .02, .05, .1]
max_ratess = [Uniform(200,400), Uniform(300,400), Uniform(150,300), Uniform(25,225),
              Uniform(20,140), Uniform(15,105), Uniform(10,108)]
required_rates = []
for nneurons, pstc, max_rates in zip(nneuronss, pstcs, max_ratess):
    rate = test_ff_dynamics_1syn(nneurons=nneurons, pstc=pstc, max_rates=max_rates)
    required_rates.append(rate)
```

The blue line shows the signal observed by the Probe, and the red line shows the predicted signal obtained by simply filtering the input $u(t)$ with the synaptic time constant. We see that the dynamics of the observed $x(t)$ matches the predicted dynamics, so we say that the network dynamics are governed solely by the output synaptic filter. 

Using the predicted and observed $x(t)$, the plots also show the RMSE and relative RMSE (RRMSE). Besides varying the synaptic time constant, we've also set the number of neurons and their firing rates so that the RRMSE is approximately constant. Plotting the required spike rates versus the synaptic time constant,

```python
def plot_tau_spikes(pstcs, required_rates):
    figure(figsize=(6,3))
    loglog(pstcs, required_rates, '-o')
    xlabel(r'$\tau_{syn}$', fontsize=18)
    ylabel('spikes/s', fontsize=16)
    title('Required spike rate for fixed RRMSE')
```

```python
plot_tau_spikes(pstcs, required_rates)
```

we see that the require spike rate drops with the synaptic time constant for fixed RRMSE. Intuitively, the longer the synaptic time constant, the cleaner the signal, so it makes sense that we should need lower spike rates for higher time constants.

# We can use NEF to control the system dynamics

Now we'll look at whether we can use NEF dynamics to meet the SNR/system time constant spec with lower firing rates. NEF uses recurrence to create the neural system

![neurally plausible dynamical system](figures/neural_system.png "neurally plausible dynamical system")

where

\begin{align}
x(t) &= h(t)*[A'x(t)+B'u(t)] \\
X(s) &= H(s)[A'X(s)+B'U(s)] \\
\end{align}

and $h(t)$ is the impulse response of the synapse. 

\begin{align}
h(t) &= \frac{1}{\tau}e^{-t/\tau}\mathbf{1}[t>=0] \\
H(s) &= \frac{1}{\tau s+1} \\
\end{align}

To implement dynamics specified by

\begin{align}
\dot{x}(t) &= Ax(t)+Bu(t) \\
sX(s) &= AX(s)+BU(s) \\
\end{align}

we use

\begin{align}
A' &= \tau A+I \\
B' &= \tau B \\
\end{align}

To implement the neural system with nengo, we use

![NEF dynamical network](figures/dyn_nengo.png "NEF dynamical network")

where the decoders are trained for identity. The input to the Ensemble as well as the instaneous value of the decoded spikes both represent $Ax(t)+Bu(t)$. For the double filtering reason described in the feedfoward cases, we do not use a synapse in the Connection between the Node and the Ensemble.

To implement a system with dynamics

\begin{align}
\tau_{sys}\dot{x}(t) &= -x(t)+u(t) \\
\end{align}

we use

\begin{align}
A' &= -\frac{\tau}{\tau_{sys}}+1 \\
B' &= \frac{\tau}{\tau_{sys}} \\
\end{align}

We use this formulation and implement a system with fixed dynamics for different synaptic time constants to answer whether using an NEF dynamical formulation can help us reduce firing rates.

```python
def build_dyn_sys(nneurons, dim, pstc, A, B, radius=1., pstc_in=None, max_rates=Uniform(200,400)):
    """Build a recurrent network implementing the specified dynamics"""
    A = pstc*A+np.eye(A.shape[0])
    B = pstc*B
    net = nengo.Network(seed=1)
    with net:
        net.input = nengo.Node(None, size_in=dim)
        net.ens = nengo.Ensemble(nneurons, dim, radius=radius, max_rates=max_rates, seed=1)
        net.output = nengo.Node(None, size_in=dim)
        nengo.Connection(net.input, net.ens, transform=B, synapse=pstc_in)
        nengo.Connection(net.ens, net.ens, transform=A, synapse=pstc)
        nengo.Connection(net.ens, net.output, synapse=pstc)
        net.probe_spikes = nengo.Probe(net.ens.neurons)
    return net, A, B
```

```python
def test_r_dynamics(nneurons, pstc, tau_sys, radius, max_rates):
    dt = .001
    A = np.array([[-1./tau_sys]])
    B = np.array([[1./tau_sys]])
    net, Ap, Bp = build_dyn_sys(nneurons=nneurons, dim=1, pstc=pstc, A=A, B=B, radius=radius, max_rates=max_rates)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=None)
        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_ens_in = nengo.Probe(net.ens, 'input')
        probe_out = nengo.Probe(net.output, synapse=None)
        probe_tau_sys = nengo.Probe(stimulus, synapse=tau_sys)
    sim = nengo.Simulator(net)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_tau_sys])
    sim_rms = rms(sim.data[probe_tau_sys])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_ens_in], c='k', alpha=.4, label='ens. input')
    plot(t, sim.data[probe_out], 'b', label='observed ' + r'$x(t)$')
    plot(t, sim.data[probe_tau_sys], 'r', label='predicted ' + r'$x(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    mean_ens_rate = np.mean(spike_dat)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (mean_ens_rate))
    fig.subplots_adjust(top=.8)
    suptitle((r"$n=%d$  $\tau_{sys}=%.2f$  $\frac{\tau_{syn}}{\tau_{sys}}=%.1f$  $A'=%.1f$  $B'=%.1f$" % 
              (nneurons, tau_sys, pstc/tau_sys, Ap[0], Bp[0])), fontsize=16)
    return mean_ens_rate
    
tau_sys = .01
nneuronss = [500, 200, 100, 100, 100, 100, 100]
pstcs = [.001, .002, .005, .01, .02, .05, .1]
radii = [1.,   1.,   1.,   1.,  1.5, 5.,  10.]
max_ratess = [
    Uniform(300, 450), Uniform(300, 450), Uniform(175, 300), Uniform(100, 175),
    Uniform(75,150), Uniform(100, 175), Uniform(100, 175)]
required_rates = []
for nneurons, pstc, radius, max_rates in zip(nneuronss, pstcs, radii, max_ratess):
    rate = test_r_dynamics(nneurons, pstc, tau_sys, radius, max_rates)
    required_rates.append(rate)
```

The network implements the desired system dynamics. Each row shows the performance of the network for a different synaptic time constant for fixed system time constant. The number of neurons and spike rates are controlled to produce approximately the same RMSE. The left column shows the network input, the input to the Ensemble, and compares the observed $x(t)$ to the predicted $x(t)$. The right column shows the Ensemble's total spiking rate.

As an aside, note how spiky the Ensemble input becomes as the system time constant drops below the synaptic time constant. To accomodate the excursions of the Ensemble input, we increase the Ensemble radius. The Ensemble input represents $A'x(t)+B'u(t)$, so the large spikes in the Ensemble's input result from a large input gain $B'=\frac{\tau}{\tau_{sys}}$. As $\tau_{sys}$ decreases relative to the synaptic $\tau$, $B'$ increases. When $B'$ is large, a sudden change in system input $u(t)$ produces a large change in the Ensemble's input until the decoded $x(t)$ can catch up. The [Appendix](#Avoid-large-input-transients-by-rescaling-the-system) describes one method to avoid the large input transients for this kind of network structure. However, the input range problem solves itself if we use a spiking input as will be seen soon.

With the RMSE controlled, we see that we can use lower spike rates with increasing synaptic time constant but _**only up to a point**_. Plotting the Ensemble firing rates required to maintain a fixed RMSE,

```python
def plot_tau_sys_spks(pstcs, tau_sys, required_rates):
    figure(figsize=(6,3))
    loglog(np.array(pstcs)/tau_sys, required_rates, '-o')
    axvline(1, c='r', alpha=.5, label=r'$\tau_{syn}=\tau_{sys}$')
    legend(loc='upper right', fontsize=16)
    xlabel(r'$\tau_{syn}/\tau_{sys}$', fontsize=18)
    ylabel('spikes/s', fontsize=16)
    title('Required spike rate for fixed RMSE')
```

```python
plot_tau_sys_spks(pstcs, tau_sys, required_rates)
```

We see that above $\tau_{syn}=\tau_{sys}$ ($|A'|>1$), there isn't an additional reduction in firing rates required to maintain the same RMSE.

# NEF system dynamics are preserved with spiking input

To make our simulation more realistic, we add an Ensemble after the input Node to convert $u(t)$ into spikes so that the recurrent portion of the Network receives spikes as input. We tap the system at the input to the Ensemble's synapse so our Probe sees $x(t)$ after filtering the decoded spikes. If we were to send the signal at the Ensemble's output (which is spikes) to a synapse and then the Probe, we would see $x(t)*h(t)$.

![NEF dynamical system with spiking input](figures/dyn_nengo_probe_tap.png "NEF dynamical system with spiking input")

```python
def build_dyn_sys_spksin(nneurons, dim, pstc, A, B, radius=1., max_rates=Uniform(200,400),
                         spk_nneurons=100, max_rates_in=Uniform(200,400)):
    """Build a recurrent network implementing the specified dynamics"""
    A = pstc*A+np.eye(A.shape[0])
    B = pstc*B
    net = nengo.Network(seed=1)
    with net:
        net.input = nengo.Node(None, size_in=dim)
        net.spike_gen = nengo.Ensemble(spk_nneurons, dim, max_rates=max_rates_in, seed=1)
        net.tap = nengo.Node(None, size_in=dim)
        net.ens = nengo.Ensemble(nneurons, dim, radius=radius, max_rates=max_rates, seed=1)
        nengo.Connection(net.input, net.spike_gen, synapse=None)
        nengo.Connection(net.spike_gen, net.tap, transform=B, synapse=None)
        nengo.Connection(net.tap, net.ens, synapse=pstc)
        nengo.Connection(net.ens, net.tap, transform=A, synapse=None)
        net.probe_spikes = nengo.Probe(net.ens.neurons)
        net.probe_input_spikes = nengo.Probe(net.spike_gen.neurons)
    return net, A, B
```

```python
def test_r_dynamics_spksin(nneurons, pstc, tau_sys, radius, max_rates, max_rates_in):
    dt = .001
    A = np.array([[-1./tau_sys]])
    B = np.array([[1./tau_sys]])
    net, Ap, Bp = build_dyn_sys_spksin(
        nneurons=nneurons, dim=1, pstc=pstc, A=A, B=B, radius=radius, 
        max_rates=max_rates, max_rates_in=max_rates_in)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=None)
        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_ens_in = nengo.Probe(net.ens, 'input')
        probe_out = nengo.Probe(net.tap, synapse=pstc)
        probe_tau_sys = nengo.Probe(stimulus, synapse=tau_sys)
    sim = nengo.Simulator(net)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_tau_sys])
    sim_rms = rms(sim.data[probe_tau_sys])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,4))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_ens_in], c='k', alpha=.4, label='ens. input')
    plot(t, sim.data[probe_out], 'b', label='observed ' + r'$x(t)$')
    plot(t, sim.data[probe_tau_sys], 'r', label='predicted ' + r'$x(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    ylabel('filtered\n spike rate (spks/s)')
    in_spike_dat = np.sum(sim.data[net.probe_input_spikes], axis=1)
    ens_spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    filtfilt(in_spike_dat, synapse=.001, dt=dt, copy=False)
    filtfilt(ens_spike_dat, synapse=.001, dt=dt, copy=False)
    net_spike_dat = ens_spike_dat + in_spike_dat
    mean_in_rate = np.mean(in_spike_dat)
    mean_ens_rate = np.mean(ens_spike_dat)
    mean_net_rate = np.mean(net_spike_dat)
    plot(t, in_spike_dat, 'k', alpha=.4, label='input spike rate')
    plot(t, ens_spike_dat, 'b', label='Ensemble spike rate')
    plot(t, net_spike_dat, 'k', label='total spike rate')
    legend(loc='lower right')
    title_str = 'mean {input, Ensemble, total} spike rate =\n{%.0f, %.0f, %.0f} (spks/s)' % (
        mean_in_rate, mean_ens_rate, mean_net_rate)
    title(title_str)
    fig.subplots_adjust(top=.8)
    suptitle((r"$n=%d$  $\tau_{sys}=%.2f$  $\frac{\tau_{syn}}{\tau_{sys}}=%.1f$  $A'=%.1f$  $B'=%.1f$" % 
              (nneurons, tau_sys, pstc/tau_sys, Ap[0], Bp[0])), fontsize=14)
    rate = mean_net_rate
    return rate
    
tau_sys = .01
nneuronss = [500, 200, 100, 1, 100, 100, 100]
pstcs = [.001, .002, .005, .01, .02, .05, .1]
radii = [1.,   1.,   1.,   1.,  1., 1.,  1.]
max_in_rates = [
    Uniform(200,400), Uniform(200,400), Uniform(200,400), Uniform(100,175),
    Uniform(200,400), Uniform(200,400), Uniform(200,400)]
max_ens_rates = [
    Uniform(300, 425), Uniform(250, 340), Uniform(100, 175), Uniform(10, 20),
    Uniform(25,120), Uniform(50, 150), Uniform(100, 170)]
required_rates = []
for nneurons, pstc, radius, max_ens_rate, max_in_rate in zip(nneuronss, pstcs, radii, max_ens_rates, max_in_rates):
    rate = test_r_dynamics_spksin(nneurons, pstc, tau_sys, radius, max_ens_rate, max_in_rate)
    required_rates.append(rate)
```

These plots are in the same style as the no input-noise simulations above. I also plot the input spike rate in the right column. 

Note that for $\tau_{syn}=\tau_{sys}$, I do not show the Ensemble spike rate. In this case, $A'=0$ so there is no feedback, and the only meaningful spike rate is the input spike rate.

Note also that we avoided the Ensemble range issue of the previous setup because the Ensemble's input is after the filter. In the no input-noise setup, the Ensemble input came before the filter.

```python
plot_tau_sys_spks(pstcs, tau_sys, required_rates)
```

Consistent with the observations of the no input-noise setup, we see that the required spike rate to maintain a fixed RMSE decreases with longer synaptic time constants but only up to when $\tau_{syn}=\tau_{sys}$.

# Noise in the feedback limits spike rate reduction

To understand why the required number of spikes doesn't decrease with increasing $\tau_{syn}$, we expand Ensembles of the system to understand where noise is coming from and how they propagate to the system state.

![Ensemble expansion](figures/ens_expansion.png "Ensemble expansion")

In the context of this notebook, Ensembles are comprised of somas and decoding weights $d$ choosen to decode identity. We would like $y(t)=x(t)$. However, somas output spike. That is, they add noise to $x(t)$. The injected spike noise is modeled by $\eta$. After the soma, the spikes are scaled by the decoding weights $d$ such that their mean value is $x(t)$ (because we are decoding identity) and passed as the output $y(t)$ of the Ensemble. Therefore, $y(t) = x(t) + d\eta$. We know that $d$ decreases in proportion to the height of the tuning curves and the number of tuning curves. That is, for a faster Ensemble spiking rate, we can scale down $d$ and reduce the amount of spike noise passed to the output. Replacing the Ensembles with their noise injection counterparts, we have

![NEF dynamical network with noise modeled](figures/dyn_nengo_noise.png "NEF dynamical network with noise modeled")

For this system,

\begin{align}
X(s) &= H(s)\left[A'(X(s)+d_2\eta_2(s))+B'(U(s)+d_1\eta_1(s))\right] \\
(1-A'H(s))X(s) &= A'H(s)d_2\eta_2(s) + B'H(s)U(s)+B'H(s)d_1\eta_1(s) \\
X(s) &= \frac{A'H(s)}{1-A'H(s)}d_2\eta_2(s) + \frac{B'H(s)}{1-A'H(s)}U(s) + \frac{B'H(s)}{1-A'H(s)}d_1\eta_1(s) \\
 &= \frac{A'}{1/H(s)-A'}d_2\eta_2(s) + \frac{B'}{1/H(s)-A'}U(s) + \frac{B'}{1/H(s)-A'}d_1\eta_1(s) \\
 &= \frac{A'}{\tau_{syn}s+1-A'}d_2\eta_2(s) + \frac{B'}{\tau_{syn}s+1-A'}U(s) + 
    \frac{B'}{\tau_{syn}s+1-A'}d_1\eta_1(s) \\
\end{align}

Where we see components for the spike noise injections $\eta_1$ and $\eta_2$ and the input $u$. Next, what is the effect of the two noise components when we've set $A'$ and $B'$ to implement a system with first order dynamics governed by $\tau_{sys}$? Recall that 

\begin{align}
A' &= -\frac{\tau_{syn}}{\tau_{sys}}+1 \\
B' &= \frac{\tau_{syn}}{\tau_{sys}} \\
\end{align}

Starting with the $\eta_1$ component,

\begin{align}
\frac{B'}{\tau_{syn}s+1-A'}d_1\eta_1(s) &= 
    \frac{\tau_{syn}/\tau_{sys}}{\tau_{syn}s+1+\tau_{syn}/\tau_{sys}-1}d_1\eta_1(s) \\
 &= \frac{\tau_{syn}/\tau_{sys}}{\tau_{syn}s+\tau_{syn}/\tau_{sys}}d_1\eta_1(s) \\
 &= \frac{1}{\tau_{sys}s+1}d_1\eta_1(s) \\
\end{align}

We see that $\eta_1$ is filtered by the system time constant but is unaffected by the synaptic time constant, which explains why adding spiking input noise did not affect the trend observed in the simulations. Moving on to $\eta_2$,

\begin{align}
\frac{A'}{\tau_{syn}s+1-A'}d_2\eta_2(s) &= 
    \frac{1-\tau_{syn}/\tau_{sys}}{\tau_{syn}s+1+\tau_{syn}/\tau_{sys}-1}d_2\eta_2(s) \\
 &= \frac{1-\tau_{syn}/\tau_{sys}}{\tau_{syn}s+\tau_{syn}/\tau_{sys}}d_2\eta_2(s) \\
 &= \frac{\tau_{sys}/\tau_{syn}-1}{\tau_{sys}s+1}d_2\eta_2(s) \\
\end{align}

We see that $\eta_2$ is likewise filtered by $\tau_{sys}$ but also scaled by a gain of $\tau_{sys}/\tau_{syn}-1$. Plotting the gain for fixed $\tau_{sys}$ as $\tau_{syn}$ varies,

```python
# plot the gain as tau_syn varies for fixed tau_sys
tau_syn = np.logspace(-1, 1) # in units of tau_sys
gain = 1./tau_syn - 1
figure(figsize=(6,3))
semilogx(tau_syn, np.abs(gain))
axvline(1, c='r', alpha=.5, label=r'$\tau_{syn}=\tau_{sys}$')
legend(loc='upper right')
xlabel(r'$\tau_{syn}/\tau_{sys}$', fontsize=18)
ylabel(r'$|\tau_{sys}/\tau_{syn}-1|$', fontsize=20);
```

we see that that the gain of the noise injected by the recurrently connected Ensemble has three regions.
 - When $\tau_{syn} < \tau_{sys}$ (synapse is faster than the system), the gain becomes very large. This explains why we needed a high spike rate for $\tau_{syn} \ll \tau_{sys}$.
 - When $\tau_{syn} = \tau_{sys}$ (synapse is the same speed as the system), the gain is $0$ (because there is no recurrent connection).
 - When $\tau_{syn} > \tau_{sys}$ (synapse is slower than the system), the gain magnitude approaches $1$. This explains why increasing $\tau_{syn}$ beyond $\tau_{sys}$ had little to no affect on the required spike rate.
 
Putting $\eta_1$ and $\eta_2$ back into the equation for $X(s)$, we have

\begin{align}
X(s) &= \frac{B'}{\tau_{syn}s+1-A'}U(s) + 
    \frac{1}{\tau_{sys}s+1}d_1\eta_1(s) + \frac{\tau_{sys}/\tau_{syn}-1}{\tau_{sys}s+1}d_2\eta_2(s) \\
\end{align}

# We can use adaptive neurons to speed up the dynamics

The adaptive neuron is being analyzed [here](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/neuron/adaptive_LIF_neuron.ipynb). 

```python
def build_aff_sys(nneurons, dim, pstc_out, tau_n, inc_n, radius=1., pstc_in=None, max_rates=Uniform(100,200)):
    neuron_type = nengo.AdaptiveLIF(tau_n=tau_n, inc_n=inc_n)
    net = nengo.Network(seed=1)
    with net:
        net.ens = nengo.Ensemble(
            nneurons, dim, radius=radius, neuron_type=neuron_type, max_rates=max_rates, seed=1)
        net.input = nengo.Node(None, size_in=dim)
        net.output = nengo.Node(None, size_in=dim)
        nengo.Connection(net.input, net.ens, synapse=pstc_in)
        nengo.Connection(net.ens, net.output, synapse=pstc_out)
        net.probe_spikes = nengo.Probe(net.ens.neurons)
    return net
```

```python
def test_aff_dynamics(nneurons, pstc, tau_n, inc_n, max_rates):
    dt = .001
    net = build_aff_sys(nneurons=100, dim=1, pstc_out=pstc, tau_n=tau_n, inc_n=inc_n, max_rates=max_rates)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=None)
        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_out = nengo.Probe(net.output, synapse=None)
        probe_pstc = nengo.Probe(stimulus, synapse=pstc)

    sim = nengo.Simulator(net, dt=dt)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_pstc])
    sim_rms = rms(sim.data[probe_pstc])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_out], 'b', label='neural ' + r'$x(t)$')
    plot(t, sim.data[probe_pstc], 'r', label=r'$h(t)*u(t)$')
    legend(loc='lower right')
#     title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    mean_rates = np.mean(spike_dat)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (mean_rates))
    fig.subplots_adjust(top=.85)
    suptitle(r'$\tau_f=%.3f$ $\tau_{sys}=%.3f$' % (tau_n, pstc), fontsize=14)
    return

nneuronss = [100, 100, 100, 100, 200, 400, 400]
pstcs = [.001, .002, .005, .01, .02, .05, .1]
tau_ns = [.001, .002, .005, .01, .02, .05, .1]
inc_ns = [.1, .2, .5, 1., 2., 5., 10.]
max_ratess = [Uniform(100,200), Uniform(100,200), Uniform(100,200), Uniform(100,200),
              Uniform(100,200), Uniform(5,50), Uniform(5,50)]
required_rates = []
for nneurons, pstc, tau_n, inc_n, max_rates in zip(nneuronss, pstcs, tau_ns, inc_ns, max_ratess):
    rate = test_aff_dynamics(nneurons=nneurons, pstc=pstc, tau_n=tau_n, inc_n=inc_n, max_rates=max_rates)
    required_rates.append(rate)
```

## A feedfoward network with two synapses double filters the input

The feedfoward network we previously [considered](#Feedfoward-dynamics-are-governed-by-the-synaptic-time-constant) only had one synapse in the Connecting from the Ensemble to the Probe. Here we consider if we also add a synapse in the Connection between the Node and the Ensemble.

![feedfoward network with input synapse](figures/feed_foward_nengo_2syn.png "feedfoward network with input synapse")

Including a synapse in the Connection between the input Node and the Ensemble causes the network to filter the input twice. That is,

\begin{align}
x(t) &= h(t) * h(t) * u(t) \\
X(s) &= H(s)H(s)U(s) \\
\end{align}

Note that this is not the same filtering with twice the synaptic time constant. For $H(s)=\frac{1}{\tau s+1}$

\begin{align}
X(s) &= \frac{1}{\tau s+1}\frac{1}{\tau s+1}U(s) \\
 &= \frac{1}{\tau^2s^2+2\tau s+1}U(s) \\
 &\neq \frac{1}{2\tau s+1}U(s) \\
\end{align}

We see that $X(s)$ is a second order system and _**not**_ a first order system with double the time constant.

```python
def test_ff_dynamics_2syn(pstc):
    dt = .001
    net = build_ff_sys(nneurons=100, dim=1, pstc_out=pstc)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=pstc)
        node1 = nengo.Node(lambda t, x:x, size_in=1)
        nengo.Connection(stimulus, node1, synapse=pstc)
        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_out = nengo.Probe(net.output, synapse=None)
        probe_pstc = nengo.Probe(node1, synapse=pstc)
    sim = nengo.Simulator(net, dt=dt)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_pstc])
    sim_rms = rms(sim.data[probe_pstc])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_out], 'b', label='observed ' + r'$x(t)$')
    plot(t, sim.data[probe_pstc], 'r', label='predicted ' + r'$x(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (np.mean(spike_dat)))
    fig.subplots_adjust(top=.85)
    suptitle(r'$\tau_{syn}=%.3f$' % pstc, fontsize=14)

pstcs = [.001, .01, .1]
for pstc in pstcs:
    test_ff_dynamics_2syn(pstc=pstc)
```

The predicted second order dynamics matches the observed $x(t)$ dynamics.

## Avoid large input transients by rescaling the system

The above one-synapse [implementation](#We-can-use-NEF-to-control-the-system-dynamics) of NEF dynamics had to cope with large transients in the input seen by the Ensemble. The problem results because $B'$ scales the input and tracks the speedup of the system. Faster systems have larger swings in the input. To avoid this problem, imagine that instead of implementing the low-pass filter

\begin{align}
\tau_{sys}\dot{x} &= -x+u \\
\dot{x} &= -\frac{1}{\tau_{sys}}x+\frac{1}{\tau_{sys}}u \\
\end{align}

we implemented 

\begin{align}
\tau_{sys}\dot{x} &= -x+u \\
\dot{x} &= -\frac{1}{\tau_{sys}}x+u \\
\end{align}

and took the output of the system to be

\begin{align}
y &= Cx \\
\end{align}

where $C=\frac{1}{\tau_{sys}}$ then

\begin{align}
\dot{y} &= C\dot{x} \\
 &= \frac{1}{\tau_{sys}}\left(-\frac{1}{\tau_{sys}}x+u\right) \\
 &= \frac{1}{\tau_{sys}}\left(-y+u\right) \\
 &= -\frac{1}{\tau_{sys}}y+\frac{1}{\tau_{sys}}u \\
\tau_{sys}\dot{y} &= -y+u \\
\end{align}

which is exactly the dynamical equation of the original low-pass filter. To implement this system using neurons, we use 

\begin{align}
A' &= -\frac{\tau}{\tau_{sys}}+1 \\
B' &= \tau \\
C &= \frac{1}{\tau_{sys}} \\
\end{align}

and build

![reshuffled NEF dynamical system](figures/dyn_nengo_reshuffled.png "reshuffled NEF dynamical system")

```python
def test_r_dynamics_reshuffled(nneurons, pstc, tau_sys, radius, max_rates):
    dt = .001
    A = np.array([[-1./tau_sys]])
    B = np.array([[1.]])
    C = np.array([[1./tau_sys]])
    net, Ap, Bp = build_dyn_sys(nneurons=nneurons, dim=1, pstc=pstc, A=A, B=B, radius=radius, max_rates=max_rates)
    with net:
        stimulus = nengo.Node(square_wave)
        nengo.Connection(stimulus, net.input, synapse=None)
        probe_in = nengo.Probe(stimulus, synapse=None)
        probe_ens_in = nengo.Probe(net.ens, 'input')
        node_out = nengo.Node(lambda t, x: x, size_in=1)
        nengo.Connection(net.output, node_out, transform=C, synapse=None)
        probe_out = nengo.Probe(node_out, synapse=None)
        probe_tau_sys = nengo.Probe(stimulus, synapse=tau_sys)
    sim = nengo.Simulator(net)
    sim.run(1, progress_bar=False)
    t = sim.trange()
    sim_rmse = rmse(sim.data[probe_out], sim.data[probe_tau_sys])
    sim_rms = rms(sim.data[probe_tau_sys])
    sim_rrmse = sim_rmse / sim_rms
    fig = figure(figsize=(16,3))
    subplot(121)
    plot(t, sim.data[probe_in], 'k', label='input ' + r'$u(t)$')
    plot(t, sim.data[probe_ens_in], c='k', alpha=.4, label='ens. input')
    plot(t, sim.data[probe_out], 'b', label='neural ' + r'$y(t)$')
    plot(t, sim.data[probe_tau_sys], 'r', label='predicted ' + r'$y(t)$')
    legend(loc='lower right')
    title(r'$RMSE=%.3f$ $RMS=%.3f$ $RRMSE=%.3f$' % (sim_rmse, sim_rms, sim_rrmse))
    subplot(122)
    spike_dat = np.sum(sim.data[net.probe_spikes], axis=1)
    filtfilt(spike_dat, synapse=.001, dt=dt, copy=False)
    plot(t, spike_dat)
    ylabel('filtered\nEns. spike rate (spks/s)')
    title('mean Ens. spike rate = %.0f (spks/s)' % (np.mean(spike_dat)))
    fig.subplots_adjust(top=.8)
    suptitle((r"$n=%d$  $\tau_{sys}=%.2f$  $\frac{\tau_{syn}}{\tau_{sys}}=%.1f$  $A'=%.1f$  $B'=%.1f$" % 
              (nneurons, tau_sys, pstc/tau_sys, Ap[0], Bp[0])), fontsize=16)
    
tau_sys = .01
nneuronss = [500, 200, 100, 100, 100, 100, 100]
pstcs = [.001, .002, .005, .01, .02, .05, .1]
radii = [.01,   .01,   .01, .01,  .03, .1,  .2]
max_ratess = [
    Uniform(300, 450), Uniform(300, 450), Uniform(175, 300), Uniform(100, 200),
    Uniform(150,250), Uniform(200, 300), Uniform(200, 325)]
for nneurons, pstc, radius, max_rates in zip(nneuronss, pstcs, radii, max_ratess):
    test_r_dynamics_reshuffled(nneurons, pstc, tau_sys, radius, max_rates)
```

In some sense, we've traded one problem for another. The Ensemble inputs are now tiny because $B'$ is small, so we reduce the radius of the Ensemble.
