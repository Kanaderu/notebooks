```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import nengo

from neuron.utils.plot import make_blue_cmap, make_color_cycle
```

```python
def square_wave_fun(period=.1, amplitude=.5):
    return lambda t: amplitude*(2.*(int(t/period) % 2) - 1.)

T = .5
N_ens = 10
bcmap = make_blue_cmap(1., 0.)
```

```python
def build_chain(N_ens=10, conn_tau=.01, probe_tau=.01, N_neurons=100):
    if isinstance(conn_tau, (float, int)) or (conn_tau is None):
        conn_tau =  [conn_tau for i in xrange(N_ens-1)]
    if isinstance(probe_tau, (float, int)) or (probe_tau is None):
        probe_tau = [probe_tau for i in xrange(N_ens)]
    assert len(probe_tau) == N_ens
    assert len(conn_tau) == N_ens
    net = nengo.Network()
    net.e = []
    net.c = []
    net.p = []
    with net:
        for i in xrange(N_ens):
            net.e.append(nengo.Ensemble(N_neurons, 1))
            net.p.append(nengo.Probe(net.e[-1], synapse=probe_tau[i]))
        for i in xrange(N_ens-1):
            net.c.append(nengo.Connection(net.e[i], net.e[i+1], synapse=conn_tau[i]))
        net.stim = nengo.Node(None, size_in=1)
        nengo.Connection(net.stim, net.e[0], synapse=None)
    return net
```

```python
def run_chain_exp(N_ens, conn_tau, probe_tau, N_neurons=100):
    net = build_chain(N_ens, conn_tau, probe_tau)
    with net:
        stim = nengo.Node(square_wave_fun(), size_out=1)
        nengo.Connection(stim, net.stim, synapse=None)
    sim = nengo.Simulator(net)
    sim.run(T, progress_bar=False)
    t = sim.trange()

    bcc = make_color_cycle(N_ens, bcmap)
    plt.figure(figsize=(8, 6))
    for p_idx, p in enumerate(net.p):
        plt.plot(t, sim.data[p], c=bcc[p_idx], alpha=.5)
    plt.xlabel('time')
    plt.ylabel('probed data')
```

```python
run_chain_exp(N_ens, .01, .01)
```

```python
run_chain_exp(N_ens, None, .01)
```
