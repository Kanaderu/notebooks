# Notebook description

The soma dynamics do not seem to affect the system dynamics.

What if we just chain a bunch of somas together without any synapses in between?

We'll see that without synaptic filtering, noise accumulates through the network. Noise alone wouldn't be a problem if  we could filter it later. Actually, filtering later with a higher order filter with be less noisy than filtering at each stage. However, noise also drives the Ensembles outside of their representation range. Outside of the representational range, neurons saturate and cause errors to accumulate in the decoded values transmitted between Ensembles.

# Setup

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from scipy.special import binom

import nengo
from nengo import LinearFilter

from neuron.utils.plot import make_blue_cmap, make_red_cmap, make_color_cycle
```

```python
def square_wave_fun(period=.1, amplitude=.5):
    return lambda t: amplitude*(2.*(int(t/period) % 2) - 1.)

T = .3
N_ens = 4
tau = .01
bcmap = make_blue_cmap(1., 0.)
rcmap = make_red_cmap(1., 0.)
```

```python
def build_chain(N_ens=4, conn_syn=.01, probe_syn=.01, N_neurons=100):
    if not isinstance(conn_syn, (list, np.ndarray)):
        conn_syn =  [conn_syn for i in xrange(N_ens-1)]
    if not isinstance(probe_syn, (list, np.ndarray)):
        probe_syn = [probe_syn for i in xrange(N_ens)]
    assert len(conn_syn) == N_ens-1
    assert len(probe_syn) == N_ens
    net = nengo.Network(seed=1)
    net.e = []
    net.c = []
    net.p = []
    net.cp = []
    with net:
        for i in xrange(N_ens):
            net.e.append(nengo.Ensemble(N_neurons, 1, radius=1., seed=1))
            net.p.append(nengo.Probe(net.e[-1], synapse=probe_syn[i]))
        for i in xrange(N_ens-1):
            net.c.append(nengo.Connection(net.e[i], net.e[i+1], synapse=conn_syn[i]))
            net.cp.append(nengo.Probe(net.c[-1], synapse=None))
        net.stim = nengo.Node(None, size_in=1)
        nengo.Connection(net.stim, net.e[0], synapse=None)
    return net


def run_chain_exp(N_ens, conn_syn, probe_syn, N_neurons=100, pb=False):
    net = build_chain(N_ens, conn_syn, probe_syn)
    with net:
        stim = nengo.Node(square_wave_fun(), size_out=1)
        nengo.Connection(stim, net.stim, synapse=None)
    sim = nengo.Simulator(net)
    sim.run(T, progress_bar=pb)
    t = sim.trange()

    bcc = make_color_cycle(N_ens, bcmap)
    rcc = make_color_cycle(N_ens-1, rcmap)
    plt.figure(figsize=(12, 4))
    plt.subplot(121)
    for p_idx, p in enumerate(net.p):
        plt.plot(t, sim.data[p], c=bcc[p_idx], alpha=.7)
    plt.xlabel('time', fontsize=16)
    plt.ylabel('probed ensemble data', fontsize=16)
    plt.subplot(122)
    for cp_idx, p in enumerate(net.cp):
        plt.plot(t, sim.data[p], c=rcc[cp_idx], alpha=.7)
    plt.xlabel('time', fontsize=16)
    plt.ylabel('probed connection data', fontsize=16)
```

# Simulations

```python
# syn in each conn
run_chain_exp(N_ens, tau, tau)
```

```python
# no syns in conns
run_chain_exp(N_ens, None, tau)
```

```python
# no syns in conns, probes with cumlative syns
probe_syn = []
for i in xrange(1, N_ens+1):
    den = [binom(i, j)*tau**j for j in xrange(i+1)][::-1]
    probe_syn.append(LinearFilter(1, den))
run_chain_exp(N_ens, None, probe_syn)
```

```python
# no syns in conns, last probe with k order syn
probe_syn = [LinearFilter(1, [tau, 1]) for i in xrange(N_ens-1)]
k = N_ens/2
den = [binom(k, j)*tau**j for j in xrange(k+1)][::-1]
probe_syn.append(LinearFilter(1, den))
run_chain_exp(N_ens, None, probe_syn)
```
