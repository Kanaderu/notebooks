# Notebook description

In NEF, the soma dynamics do not seem to affect the system dynamics. Intuitively, this is because of the heterogeneity of the neural activity across the input space and overall high firing rates. No matter the input value to an ensemble of neurons, there will be some neurons that are firing quickly. This means that if we're observing the output spikes of the ensemble, we needn't wait long before seeing a change in the output and inferring a change in the input.

If the soma dynamics don't affect the dynamics of the network, why don't we just chain a bunch of somas together without any synapses in between if we want to build a superfast network?

![Network of a chain of Ensembles](figures/chain.png "Network of a chain of Ensembles")

We'll see that without synaptic filtering, correlated spiking noise accumulates as the signal propogates through the network. Uncorrelated noise wouldn't necessarily be a problem if we could filter it later. Actually, filtering later with a higher order filter with be less noisy than filtering at each stage. 

However, because the progated noise is correlated, it drives the ensembles outside of their representation range. Outside of their range, neurons saturate and cause errors to accumulate in the decoded values transmitted between Ensembles.

It's important to note that this phenomenon is a matter of degree. As the input signal to an ensemble gets less noisy and less correlated, the output spikes of the ensemble will likewise be less correlated.

# Synaptic filtering between Ensembles

First, we'll simulate a chain of $4$ Ensembles with $10$ms synaptic filtering between Ensembles to establish the baseline behavior for comparison. Between each Ensemble, we decode identity.

```python
# syn in each conn
net, sim = run_chain_exp(T=.3, N_ens=4, conn_syn=.01, probe_syn=.01)
plot_decode_input(net=net, sim=sim);
```

On the left are the decoded outputs of the Ensembles with $10$ms filtering with blue being the first Ensemble and black being the last. On the right are the inputs (or outputs of the connections) seen by Ensembles with red being the input to the first Ensemble and black being the last. Because the decoded output probe filtering matches the connection synapse filtering, the first three blue lines match the last three red lines.

As expected, the signal gets progressively filtered (i.e. phase shifted) as it passes through the network. Beyond the noise added by the first Ensemble, the noise doesn't appear to grow as the signal passes through the network.

# No synaptic filtering between Ensembles

Now, we run the same $4$ Ensemble network but without any synaptic filtering between Ensembles. We're still probing the decoded outputs of the Ensembles with $10$ms filters.

```python
# no syns in conns
net, sim = run_chain_exp(T=.3, N_ens=4, conn_syn=None, probe_syn=.01)
plot_decode_input(net=net, sim=sim);
```

Here, we see that there is no phase shifting between the decoded outputs of the Ensembles. This is encouraging as it confirms our intuition that by connecting the somas without any intervening synapses, we can pass the signal through the network very quickly. Unfortunately, we also see that the noise accumulates rapidly and causes the signal to slowly drift as it passes through the Ensembles. We can also zoom in time and look more closely at the decodes and Ensemble inputs.

```python
# no syns in conns, zoomed
net, sim = run_chain_exp(T=.3, N_ens=4, conn_syn=None, probe_syn=.01, probe_spikes=True)
fig, axs = plot_decode_input(net=net, sim=sim)
axs[0].set_xlim(.13, .15);
```

Zooming in, we can see that the input to each Ensemble becomes increasingly noisy as the signal passes through the Ensembles. If we look at the spike rasters, we can see why.

```python
fig, axs = plt.subplots(nrows=4, sharex=True, figsize=(12,10))
for i in xrange(4):
    rasterplot(sim.trange(), sim.data[net.sp[i]], axs[i], colors=[(0,0,0) for j in xrange(100)])
    axs[i].set_ylabel('Ensemble %d' % i, fontsize=12)
axs[-1].set_xlabel('time', fontsize=16);
```

The spike rasters show that the neurons in subsequent Ensembles become increasingly synchronized as the signal passes through the network. This happens because neurons with similar parameters and similar encoding vectors will tend to spike at the same time given noise in the input. Consider the LIF soma equation

$$\tau x = -x + u$$

If we are only feeding spikes into the soma and it has no bias, then the soma will only spike when an input spike arrives.

# Weak fast inhibition

What if we put in some weak mutual inhibition between neurons with the same encoders?

```python
net = build_chain(N_ens=4, conn_syn=None, N_neurons=64, probe_spikes=True)
with net:
    stim = nengo.Node(square_wave_fun(), size_out=1)
    nengo.Connection(stim, net.stim, synapse=None)
for ens in net.e:
    ens.encoders = np.ones((64,1))
    ens.encoders[:32] = -1.
    with net:
        inhib_mat = -np.dot(ens.encoders, ens.encoders.T).clip(0, .0001)
        nengo.Connection(ens.neurons, ens.neurons, transform=inhib_mat, synapse=.0001)
sim = nengo.Simulator(net, dt=1e-4)
sim.run(.3, progress_bar=False)
fig, axs = plot_decode_input(net=net, sim=sim)

fig, axs = plt.subplots(nrows=4, sharex=True, figsize=(12,10))
for i in xrange(4):
    rasterplot(sim.trange(), sim.data[net.sp[i]], axs[i], colors=[(0,0,0) for j in xrange(100)])
    axs[i].set_ylabel('Ensemble %d' % i, fontsize=12)
axs[-1].set_xlabel('time', fontsize=16);
```

This doesn't seem to fare much better. If the mutual inhibition is too strong, then the strongest neuron just tends to dominate. A better idea might be to inhibit based on similarity of overall tuning curves.

# No synapse in connection, equivalent synapse in probe

We can see how the decodes actually drift by using the equivalent filtering in the probes used to readout the decode.

```python
# no syns in conns, probes with cumlative syns
probe_syn = []
for i in xrange(1, 4+1):
    den = [binom(i, j)*.01**j for j in xrange(i+1)][::-1]
    probe_syn.append(LinearFilter(1, den))
net, sim = run_chain_exp(T=.3, N_ens=4, conn_syn=None, probe_syn=probe_syn)
fig, axs = plot_decode_input(net=net, sim=sim)
```

See how the filtered decode, while smooth, has drifted from a filtered version of the original input. 

If it weren't for the drift, we could just use a higher order filter at the output of the chain.

```python
# no syns in conns, last probe with k order syn
probe_syn = [LinearFilter(1, [.01, 1]) for i in xrange(4-1)]
k = 4/2
den = [binom(k, j)*.01**j for j in xrange(k+1)][::-1]
probe_syn.append(LinearFilter(1, den))
net, sim = run_chain_exp(T=.3, N_ens=4, conn_syn=None, probe_syn=probe_syn)
fig, axs = plot_decode_input(net=net, sim=sim)
```
