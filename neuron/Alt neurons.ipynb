# Simulations

```python
%pylab inline
import numpy as np
```

```python
# define models to simulate
class Synapse(object):
    def __init__(self, dt=.001, tau=.01, w=1.):
        self.dt = dt
        self.tau = tau
        self.w = w
        
        self.state = 0.
        self.a = np.exp(-self.dt/self.tau) # decay
        self.u = 1./self.dt # input unit
        
    def reset(self):
        self.state = 0.
        
    def step(self):
        self.state = self.a*self.state
        
    def spike_in(self):
        self.state += self.w*(1.-self.a)*self.u
        
class BaseNeuron(object):
    def __init__(self, dt):
        self.dt = dt
        
class ThresholdNeuron(BaseNeuron):
    """Threshold neuron fires uniformly when its synaptic input is within a specified range.
    
    Range specified by t_low and t_high
    
    Parameters
    ----------
    dt : float
        timestep of neuron
    t_low : float
        lower threshold of neuron
    t_high : float
        upper threshold of neuron
    period: int
        period of uniform firing rate in units of dt
    """
    def __init__(self, dt, t_low, t_high, period, tau_syn=0.01, w_syn=1.):
        assert t_low < t_high, "lower threshold must be less than upper threshold"
        assert isinstance(period, int), "period must be in units of dt"
        assert period > 0
        
        super(ThresholdNeuron, self).__init__(dt)
        self.t_low = t_low
        self.t_high = t_high
        self.T = period
        self.syn = Synapse(dt, tau_syn, w_syn)
        self.ctr = self.T
    
    def reset(self):
        self.ctr = self.T
        self.syn.reset()
        
    def step(self, spikes_in=0, direct_in=0.):
        self.syn.step()
        for i in xrange(spikes_in):
            self.syn.spike_in()
        in_cur = direct_in + self.syn.state

        if (in_cur >= self.t_low and in_cur <= self.t_high):
            self.ctr -= 1
            if self.ctr == 0:
                self.ctr = self.T
                return 1
            
        return 0
```

```python
# simulate neuron and synapse together
dt = .001
N = 10000

tau = .01
period = 5

t_low = 800
t_high = 1200
w = 1.

neuron = ThresholdNeuron(dt, t_low, t_high, period, tau, w)

input_rates = np.linspace(0., t_high*2., 30)
# input_rates = [100.]
spikes = np.zeros((N, len(input_rates)))
syn_states = np.zeros((N, len(input_rates)))
neu_states = np.zeros((N, len(input_rates)))

for f_idx, f_in in enumerate(input_rates):
    lam = f_in * dt
    neuron.reset()
    for t_idx in xrange(N):
        spks = np.random.poisson(lam)
        spiked = neuron.step(spikes_in=spks)
        if spiked:
            spikes[t_idx, f_idx] = 1
        syn_states[t_idx] = neuron.syn.state
        neu_states[t_idx] = neuron.ctr

spike_rates = np.sum(spikes, axis=0) / (float(N)*dt)

# figure(figsize=(8,6))
# t = dt*np.arange(N)
# subplot(211)
# plot(t, syn_states, '-')
# subplot(212)
# plot(t, neu_states, '-')


figure(figsize=(8,6))
plot(input_rates, spike_rates, '-o')
xlabel('input rate (Hz)')
ylabel('spike rate (Hz)')
# title('Neuron outputs a constant spike rate when input is within specified range')
```

```python
# validate neuron
dt = .001
N = 1000

t_low = 950.
t_high = 1050.
period = 5

neuron = ThresholdNeuron(dt , t_low, t_high, period)

inputs = np.linspace(900, 1100, 100)
spiked = np.zeros((N, len(inputs)))
for d_idx, d_in in enumerate(inputs):
    neuron.reset()
    for idx in xrange(N):
        if neuron.step(direct_in=d_in):
            spiked[idx, d_idx] = 1

spike_rates = np.sum(spiked, axis=0) / (float(N)*dt)

figure(figsize=(8,6))
plot(inputs, spike_rates, '-o')
xlabel('input current')
ylabel('spike rate (Hz)')
title('Neuron outputs a constant spike rate when input is within specified range');
```

```python
# validate synapse
dt = .001
N = 1000 # time steps to simulate
f = 1000. # input spike rate

tau = .01
w = 1.
syn = Synapse(dt, tau, w)

lam = dt*f
syn_states = np.zeros(N)
for idx in xrange(N):
    syn.step()
    spks = np.random.poisson(lam)
    for i in xrange(spks):
        syn.spike_in()
    syn_states[idx] = syn.state

t = dt*np.arange(N)
figure(figsize=(8,6))
plot(t, syn_states, '-')
ss_mean = np.mean(syn_states[t>5.*tau])
axhline(ss_mean, color='r', label='ss mean')
axhline(f, color='g', label='input rate')
legend(loc='lower left')
title('Synapse steady state mean closely matches input frequency')
xlabel('time (s)')
ylabel('synapse state');
```
