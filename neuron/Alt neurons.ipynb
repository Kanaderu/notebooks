```python
%pylab inline
import numpy as np
```

```python
# define models to simulate
class Synapse(object):
    def __init__(self, dt=.001, tau=.01, w=1.):
        self.dt = dt
        self.tau = tau
        self.w = w
        
        self.state = 0.
        self.a = np.exp(-self.dt/self.tau) # decay
        self.u = 1./self.dt # input unit
        
    def reset(self):
        self.state = 0.
        
    def step(self):
        self.state = self.a*self.state
        
    def spike_in(self):
        self.state += self.w*(1.-self.a)*self.u
        
class BaseNeuron(object):
    def __init__(self, dt):
        self.dt = dt
        
class ThresholdNeuron(BaseNeuron):
    """Threshold neuron fires uniformly when its synaptic input is within a specified range.
    
    Range specified by t_low and t_high
    
    Parameters
    ----------
    dt : float
        timestep of neuron
    t_low : float or None
        lower threshold of neuron
    t_high : float or None
        upper threshold of neuron
    period: int
        period of uniform firing rate in units of dt
    """
    def __init__(self, dt, t_low, t_high, period, tau_syn=0.01, w_syn=1.):
        if t_low and t_high:
            assert t_low < t_high, "lower threshold must be less than upper threshold"
        assert isinstance(period, int), "period must be in units of dt"
        assert period > 0
        
        super(ThresholdNeuron, self).__init__(dt)
        self.t_low = t_low
        self.t_high = t_high
        self.T = period
        self.syn = Synapse(dt, tau_syn, w_syn)
        self.ctr = self.T
    
    def reset(self):
        self.ctr = self.T
        self.syn.reset()
    
    def _in_range(self, in_cur):
        in_range = True
        if self.t_low and in_cur < self.t_low:
            in_range = False
        if self.t_high and in_cur > self.t_high:
            in_range = False
        return in_range
        
    def step(self, spikes_in=0, direct_in=0.):
        self.syn.step()
        for i in xrange(spikes_in):
            self.syn.spike_in()
        in_cur = direct_in + self.syn.state

        if (self._in_range(in_cur)):
            self.ctr -= 1
            if self.ctr == 0:
                self.ctr = self.T
                return 1
            
        return 0
```

```python
# utilities for collecting data
def get_tuning(neuron, input_rates, n_steps):
    dt = neuron.dt
    n_inputs = len(input_rates)
    spikes = np.zeros((n_steps, n_inputs))
    for f_idx, f_in in enumerate(input_rates):
        lam = f_in * dt
        neuron.reset()
        for t_idx in xrange(n_steps):
            spks = np.random.poisson(lam)
            spiked = neuron.step(spikes_in=spks)
            if spiked:
                spikes[t_idx, f_idx] = 1

    spike_rates = np.sum(spikes, axis=0) / (float(n_steps)*dt)
    return spike_rates
```

```python
# validate synapse
dt = .001
N = 10000 # time steps to simulate
f = 1000. # input spike rate

tau = .01
w = 1.
syn = Synapse(dt, tau, w)

lam = dt*f
syn_states = np.zeros(N)
for idx in xrange(N):
    syn.step()
    spks = np.random.poisson(lam)
    for i in xrange(spks):
        syn.spike_in()
    syn_states[idx] = syn.state

t = dt*np.arange(N)
syn_ss = syn_states[t>5.*tau]
ss_mean = np.mean(syn_ss)
ss_std = np.std(syn_ss)

figure(figsize=(8,6))
plot(t, syn_states, '-')
axhline(ss_mean, color='r', linewidth=2, label='steady state mean')
axhline(f, color='g', linewidth=2, linestyle=':', label='input spike rate')
legend(loc='lower left')
xlabel('time (s)', fontsize=18)
ylabel('synapse state', fontsize=18)
title('The exponential synapse is noisy', fontsize=18);

figure(figsize=(8,6))
hist(syn_ss, bins=100, normed=True)
axvline(ss_mean, color='r', linewidth=2, label='observed ss mean')
axvline(f, color='g', linewidth=2, linestyle=':', label='input spike rate')
xlabel('synapse state', fontsize=18)
ylabel('P(syn)=synapse state', fontsize=18)
title('The input spike rate determines\nthe mean and variance of the synapse', fontsize=18);
```

```python
# validate neuron
dt = .001
N = 1000

t_low = 950.
t_high = 1050.
period = 5

neuron = ThresholdNeuron(dt , t_low, t_high, period)

inputs = np.linspace(900, 1100, 100)
spiked = np.zeros((N, len(inputs)))
for d_idx, d_in in enumerate(inputs):
    neuron.reset()
    for idx in xrange(N):
        if neuron.step(direct_in=d_in):
            spiked[idx, d_idx] = 1

spike_rates = np.sum(spiked, axis=0) / (float(N)*dt)

figure(figsize=(8,6))
plot(inputs, spike_rates, '-o')
xlabel('input current', fontsize=18)
ylabel('spike rate (Hz)', fontsize=18)
title('The threshold neuron outputs a constant spike rate in response\n' +
      'to a constant input within a specified range', fontsize=18);
```

```python
# simulate neuron and synapse together
dt = .001
n_steps = 10000
n_inputs = 30

tau = .01
period = 5

t_low = 600
t_high = 1400
w = 1.

neuron = ThresholdNeuron(dt, t_low, t_high, period, tau, w)
input_rates = np.linspace(0., t_high*2., n_inputs)
spike_rates = get_tuning(neuron, input_rates, n_steps)

th_spike_rates = np.zeros(n_inputs)
idx = np.logical_and(input_rates>=t_low, input_rates<=t_high)
th_spike_rates[idx]=1./(dt*period)

figure(figsize=(8,6))
plot(input_rates, spike_rates, '-o', label='observed spike rate')
plot(input_rates, th_spike_rates, 'r', label='constant input spike rate')
ylim(ylim()[0], 1.01/(dt*period))
legend(loc='upper right')
xlabel('input rate (Hz)', fontsize=18)
ylabel('spike rate (Hz)', fontsize=18)
title("The noisy synapse smooths out the neuron's tuning curve\n", fontsize=18);
```

```python
# simulate neuron and synapse together
figure(figsize=(16,6))

dt = .001
n_steps = 10000
n_inputs = 30

tau = .01
period = 5

t_low = 600
t_high = 1400
w = 1.

input_rates = np.linspace(0., t_high*2., n_inputs)

neuron = ThresholdNeuron(dt, t_low, None, period, tau, w)
spike_rates = get_tuning(neuron, input_rates, n_steps)
th_spike_rates = np.zeros(n_inputs)
idx = input_rates>=t_low
th_spike_rates[idx]=1./(dt*period)

subplot(121)
plot(input_rates, spike_rates, '-o', label='observed spike rate')
plot(input_rates, th_spike_rates, 'r', label='constant input spike rate')
ylim(ylim()[0], 1.01/(dt*period))
legend(loc='lower right')
xlabel('input rate (Hz)', fontsize=18)
ylabel('spike rate (Hz)', fontsize=18)
title('with no upper threshold', fontsize=18)

neuron = ThresholdNeuron(dt, None, t_high, period, tau, w)
spike_rates = get_tuning(neuron, input_rates, n_steps)
th_spike_rates = np.zeros(n_inputs)
idx = input_rates<=t_high
th_spike_rates[idx]=1./(dt*period)

subplot(122)
plot(input_rates, spike_rates, '-o', label='observed spike rate')
plot(input_rates, th_spike_rates, 'r', label='constant input spike rate')
ylim(ylim()[0], 1.01/(dt*period))
xlabel('input rate (Hz)', fontsize=18)
title('with no lower threshold', fontsize=18)

suptitle("The threshold neuron tuning can saturate", fontsize=18);
```
