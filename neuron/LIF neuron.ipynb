# Notebook description

This notebook looks at the LIF neuron. The dynamics of the LIF neuron are described by

$$\tau\dot{x}=-x+u$$

where 
 - $x$ is the neuron state
 - $\tau$ is the membrane time constant and
 - and $u$ is the input to the neuron.

When $x$ reaches a predefined threshold $x_{t}$, we declare the neuron as having spiked, reset $x$ to $0$, and hold $x$ fixed for a refactory period $t_{ref}$. 

# Summary

The firing rate will be given by

$$
f = \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u}\right)}.
$$


We can solve for the input that yields a particular firing rate

$$
u=\frac{x_t}{1-\exp\left(\frac{1}{\tau}\left(t_{ref}-\frac{1}{f}\right)\right)}
$$

# LIF neuron analysis

The output of the LIF neuron is spikes. We derive the spike rate of an LIF neuron assuming that the input is constant by finding the time it takes for the neuron to reach threshold, $t_t$, and then substitute that into the equation describing the firing rate,

$$f=\frac{1}{t_{ref}+t_t}.$$

Assuming that $x=x_0$ at $t=0$, 

\begin{align}
\tau\frac{dx}{dt} &= -x + u \\
\frac{\tau}{-x+u}dx &= dt \\
\int_{x_0}^{x}\frac{\tau}{-x+u}dx &= \int_{0}^{t} dt \\
\tau\int_{x_0}^{x}\frac{1}{-x+u}dx &= \int_{0}^{t} dt. \\
\end{align}

Substituting with $v=-x+u$ (so $-dv=dx$),

\begin{align}
-\tau\int\frac{1}{v}dv &= \int_{0}^{t} dt \\
-\tau[\ln(v)]_{v_0}^{v} &= t \\
-\tau[\ln(-x+u)]_{x_0}^{x} &= t \\
-\tau[\ln(-x+u)-\ln(-x_0+u)] &= t \\
-\tau\ln\left(\frac{-x+u}{-x_0+u}\right) &= t. \\
\end{align}

Setting $t=t_t$ when $x=x_t$, we have

$$
-\tau\ln\left(\frac{-x_t+u}{-x_0+u}\right) = t_t,
$$

so the firing rate for an LIF neuron with constant input will be

$$
f = \frac{1}{t_{ref}-\tau\ln\left(\frac{u-x_t}{u-x_0}\right)}.
$$

If we set $x_0=0$, we have the firing rate function typically seen in the literature,

$$
f = \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u}\right)}.
$$

Solving for $x$ in terms of $t$, 

\begin{align}
-\tau\ln\left(\frac{-x+u}{-x_0+u}\right) &= t \\
\frac{-x+u}{-x_0+u} &= e^{-t/\tau} \\
-x+u &= (-x_0+u)e^{-t/\tau} \\
x &= -(-x_0+u)e^{-t/\tau} + u \\
x &= x_0e^{-t/\tau}+(1-e^{-t/\tau})u. \\
\end{align}

# Theoretical plots

```python
%pylab inline
import numpy as np
```

```python
def fi(x_t, tau, tref, u):
    if isinstance(u, (float, int)):
        u = array([u], dtype=float)
    f = zeros(len(u))
    idx = u>x_t
    f[idx] = (tref-tau*log(1.-x_t/u[idx]))**-1.
    return f

def uf(x_t, tau, tref, f):
    if isinstance(f, (float, int)):
        f = array([f], dtype=float)
    u = array([nan]*len(f))
    idx = f>0
    u[idx] = x_t/(1.-exp(1./tau*(tref-1./f[idx])))
    return u
```

```python
tau = .01
tref = .001
x_t = 1.
u = linspace(0, 10, 100)

figure(figsize=(16, 6))
ax=subplot(121)
taus = logspace(-3, -1, 3)
for tau in taus:
    plot(u, fi(x_t, tau, tref, u));
legend(taus, loc='upper left')
xlabel(r'$u$', fontsize=18)
ylabel(r'$f$', fontsize=18)
title(r'vary  $\tau$,  fix  $t_{ref}=%.3f$' % (tref), fontsize=18)

tau = .01
subplot(122, sharex=ax, sharey=ax)
trefs = logspace(-4, -2, 3)
for tref in trefs:
    plot(u, fi(x_t, tau, tref, u));
legend(trefs, loc='upper left')
xlabel(r'$u$', fontsize=18)
title(r'vary  $t_{ref}$,  fix  $\tau=%.3f$' % (tau), fontsize=18);
```
