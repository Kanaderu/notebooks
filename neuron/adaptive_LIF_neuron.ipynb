# Background

An adaptive LIF neuron is a modified version of an LIF neuron ([here](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/neuron/LIF_neuron.ipynb) is an LIF neuron analysis).

The adaptive LIF neuron is goverened by

$$\tau\dot{x}=-x+u_{in}-u_f$$

where 
 - $x$ is the neuron state
 - $\tau$ is the membrane time constant and
 - $u_{in}$ is the input to the neuron
 - $u_f$ is the feedback

The dynamics of an adaptive LIF neuron are identical to an LIF neuron. However, unlike the LIF neuron, the adaptive neuron includes a feedback term.

$u_{in}$ can either be controlled directly as an input current or derived from the state of an attached synapse. If $u_{in}$ is derived from a synapse, $u_{in}=\alpha_{syn} x_{syn}$, where $x_{syn}$ is the synaptic state and $\alpha_{syn}$ scales the synaptic state. 

$u_f$ is modeled by an inhibitory synapse that receives a spike whenever the neuron spikes. Therefore, $u_{f}=\alpha_f x_f$ where $x_f$ is the feedback synaptic state and $\alpha_{f}$ scales the feedback synaptic state.

Expanding $u_{in}$ (as a synapse) and $u_f$,

$$\tau\dot{x}=-x+\alpha_{syn}x_{syn}-\alpha_f x_f$$



As with the LIF neuron, the spiking mechanism is triggered by threshold. That is,

$$
x\ge x_t \rightarrow 
  \begin{cases}
    \text{soma outputs a spike} \\
    x :=0 & t_{spk}\le t < t_{spk}+t_{ref} \\
    x_{f}\ := x_{f}+\delta(t_{spk})
  \end{cases}
$$

where
 - $x_t$ is the threshold
 - $t_{spk}$ is the time at which $x(t)$ crosses threshold and
 - $t_{ref}$ is the refractory period
 
In words, when $x$ crosses threshold $x_{t}$, we declare the neuron as having spiked, reset and hold $x$ to $0$ for a refactory period $t_{ref}$, and feed an inhibitory spike back to the adaptive feedback synapse.

# Adaptation speeds up response

We know from the synapse demo [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/synapse/synapse_demo.ipynb) that the time course of a synapse is wholly determined by its time constant and the input firing rate. The expected synapse value will approach the input firing rate exponentially at a rate set by its time constant. The shorter the synaptic time constant, the faster the expected value of the synapse will reach its steady state value.

In addition, the synaptic noise is determined by the input firing rate, the input spike time distribution, and the synaptic time constant. The shorter the synaptic time constant, the noisier the synapse.

Here we see the tradeoff of speed vs accuracy in the synaptic time constant.
 - Shorter synaptic time constants make for faster, noisier synapses.
 - Longer synaptic time constants make for slower, cleaner, synapses.

Are we doomed to live with this tradeoff? No - one thing we didn't mention is changing the input firing rate over time. Imagine that we intend on driving the synapse to a target value. Now, instead of just giving the synapse the target value as input, we initially drive the synapse with a stronger input and then back the input off as the synapse approaches the target value.

```python
# compare saturating exponential
t = np.linspace(0,8)
x1 = 1.*(1-np.exp(-t))
x2 = 2.*(1-np.exp(-t))

fig, ax = plot_continuous(t, x2, plotp={'linewidth':2, 'color':'r', 'label':'input=2'}, figp={'figsize':(8,4)})
fig, ax = plot_continuous(t, x1, plotp={'linewidth':2, 'color':'b', 'label':'input=1'}, ax=ax, ylim=(0,2.01),
                          axhline=1, axhlinep={'ls':':'},
                          legendp={'loc':'lower right'},
                          xlabel=r'time ($\tau$)', xlabelp={'fontsize':20},
                          ylabel='synapse value', ylabelp={'fontsize':16},
                          title='A stronger input pushes the synapse faster', titlep={'fontsize':16})
```

The red line reaches the blue line's steady state value faster than the blue line. What if we could initially follow the red line but then switch to the blue line? Adaptive neurons do exactly this by firing quickly initially and then slowing down to a steady state firing rate.

```python
# demo adaptive neuron
T = .2
dt = .0001
tau_m = .01
tref = .005
xt = 1.
af = 5e-2
tau_f = 1e-1

ytop = xt*1.2  # top of plot

# timeseries for a constant input
u_vals = [1.3, 2.5]
fig = None
subplots = [(2, 2, 1), (2, 2, 3), (2, 2, 2), (2, 2, 4)]
for idx, u_val in enumerate(u_vals):
    nsteps = int(np.ceil(T/dt))
    t = np.arange(nsteps)*dt
    u = np.zeros(nsteps)+u_val
    lif_spk_times, lif_x = run_lifsoma(dt, u, tau_m, tref, xt, ret_state=True)
    alif_spk_times, alif_x, alif_xf = run_alifsoma(dt, u, tau_m, tref, xt, af=af, tauf=tau_f,
                                                   ret_state=True, ret_fstate=True)

    if fig is None:
        fig, ax = plot_continuous(t, lif_x, plotp={'c':'k', 'lw':1, 'ls':':', 'label':'nonadaptive LIF state'},
                                  figp={'figsize':(16,10)}, subplotp=subplots[2*idx])
    else:
        fig, ax = plot_continuous(t, lif_x, plotp={'c':'k', 'lw':1, 'ls':':', 'label':'nonadaptive LIF state'}, fig=fig, 
                                  subplotp=subplots[2*idx])

    plot_continuous(t, alif_x, plotp={'color':'b','linewidth':2, 'label':'adaptive LIF state'}, ax=ax,
                    axhline=xt, axhlinep={'color':'r', 'ls':':'},
                    legendp={'loc':'lower right'},
                    ylabel=r'$x$', ylabelp={'fontsize':24},
                    title=r'$u_{in}=%.2f$' % u_val, titlep={'fontsize':20})
    ax.set_ylim(ax.get_ylim()[0], ytop)

    fig, ax = plot_continuous(t, alif_xf, plotp={'color':'b'}, fig=fig, subplotp=subplots[2*idx+1],
                              xlabel=r'$t$ (s)', xlabelp={'fontsize':20},
                              ylabel=r'$x_f$', ylabelp={'fontsize':24},)
fig.suptitle('Adaptation slows down the initial firing rate', fontsize=20);
```

In the upper plots, the black (blue) traces show the state of nonadaptive (adaptive) LIF neurons in response to a constant input. The lower plots show the adaptive feedback as the adaptive neuron begins to spike.

We can see that the interspike intervals increase as adaptation ($x_f$) kicks in and that the adaptive feedback behaves like like a synapse. In fact, if the adaptive neuron projects to a synapse (that has no other inputs) with the same time constant as the adaptation, the synapse will exactly match the adative feedback.

This behavior suggests a strategy for estimating when the adaptive neuron has reached steady state. First, let's pretend we knew the adaptive neuron's steady state firing rate (which is the next [section](#Steady-state-firing-rate) anyways). We know from the synapse demo [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/synapse/synapse_demo.ipynb) that a synapse's expected state approaches the input spike rate exponentially with the synaptic time constant. As adaptation behaves identically to a synapse, we know that the expected adaptation state approaches the neuron's firing rate exponentially with the adaptive feedback time constant. We also know that adaptation decreases the neuron's firing rate, so the neuron initially fires faster than at steady state. Therefore, the exepcted adaptive state approaches the steady state firing rate faster than the expected state of a hypothetical synapse receiving spikes at the steady state rate. This means that we need observe the synapse for a length of time determined by the adaptive feedback time constant. More concretely, running the adaptive neuron for something like $5\tau_f$ should suffice for declaring that the adaptation, and hence the neuron, has reached steady state. Since the firing rate is uniform at steady state, we need only measure a single interspike interval to determine the spike rate.

# Steady state firing rate

The firing rate for an LIF neuron without adaptation and with constant input is

$$
f = \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}}\right)} \qquad u_{in}>x_t
$$

To approximate the steady-state firing rate of an adaptive neuron, we first assume that the adaptive neuron behaves like the nonadaptive neuron in response to constant $u_{in}$ and $u_f$. We also assume that the constant feedback value is its steady state expected value, $u_f=E[u_f]$. Since $u_f$ is derived from a synapse, $E[u_f]=\alpha E[x_f]=\alpha f$ (see [synapse demo](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/synapse/synapse_demo.ipynb) for $E[x_f]=f$). With these approximations,

\begin{align}
f &\approx \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}-u_f}\right)} & u_{in}-u_f>x_t &,
    \quad \text{assuming $u_{in}$ and $u_f$ are constant} \\
 &\approx \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}-E[u_f]}\right)} & u_{in}-E[u_f]>x_t &,
     \quad \text{assuming }u_f=E[u_f] \\
 &= \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}-\alpha_f f}\right)} & u_{in}-\alpha f>x_t &,
     \quad E[u_f]=\alpha f \\
\end{align}


While we can't rearrange this equation to solve for $f$, we can linearly approximate the solution or find the solution numerically. First, though, let's visualize the system of equations.

```python
tau_m = .01
tref = .005
xt = 1.
af = 1e-2

th_max_f = 1./tref  # theoretical maximum firing rate
max_u = 5.  # maximum u_in to consider
max_f = th_lif_fi(max_u, tau_m, tref, xt)
f = np.linspace(5., max_f, 19)
u = th_lif_if(f, tau_m, tref, xt)  # linear in firing rate

cc = [(float(i)/len(u), 0, 0) for i in xrange(len(u))]
fig = figure(figsize=(16,6))

ax = fig.add_subplot(121)
_u = np.linspace(0, max(u), 100)
ax.plot(_u, th_lif_fi(_u, tau_m, tref, xt), 'k', linewidth=2)
ax.set_color_cycle(cc)
for idx, u_val in enumerate(u[u>xt]):
    f = th_lif_fi(u_val, tau_m, tref, xt)
    ax.plot([u_val, u[-1]], [f, f], c=cc[idx])
    ax.plot([u_val, u_val], [0, f], c=cc[idx], ls=':')
ax.set_ylim(0, th_max_f)
ax.set_xlim(0, max_u*1.001)
ax.set_xlabel(r'$u_{in}$', fontsize=20)
ax.set_ylabel(r'$f(u_{in})$ (spks / s)', fontsize=20)
ax.set_title('Nonadaptive Neuron ' + r'$(u_f=0)$', fontsize=16)

f = np.linspace(0, th_max_f, 201)
uf = af*f

ax = fig.add_subplot(122, color_cycle=cc)
for idx, u_val in enumerate(u):
    net_u = u_val - uf
    f_ss = th_lif_fi(net_u, tau_m, tref, xt)
    line = ax.plot(uf, f_ss)[0]
    if u_val in (u[0], u[-1]):
        line.set_label(r'$u_{in}=%.2f$' % u_val)
ax.plot(uf, f, 'b', lw=2, label=r'$u_f=\alpha_f f$')
ax.legend(loc='upper left', fontsize=16)
ax.set_xlabel(r'$u_f$', fontsize=20)
ax.set_ylabel(r'$f(u_f;\ u_{in})$ (spks / s)', fontsize=20)
ax.set_title('Adaptatation reduces the firing rate (' + r'$\alpha_f=%.1e$)' % (af), fontsize=16);
```

The left plot shows the usual nonadaptive LIF firing rate equation. The right plot shows how adaptive feedback ($u_f$) reduces the firing rate for fixed $u_{in}$. Every point $(u_{in}, f(u_{in}))$ on the left plot corresponds to a point $(0, f(u_{in},0))$ on the right plot's y-axis, which describes the firing rate of a neuron without any adaptation $(u_f=0)$. Further, on the right plot,
 - the firing rate decreases as inhibitory feedback increases, and
 - inhibitory feedback increases linearly with firing rate ($u_f=\alpha_ff$).

The steady state firing rate is the intersection between the $f(u_{in}, u_f)$ curve and the $u_f=\alpha_f f$ line. To see why, consider that 
 - if we decrement $u_f$ from the intersection point (i.e. move left along a fixed $u_{in}$ curve), the firing rate increases, which increases $u_f$, which cancels our initial decrement. Similarly, 
 - if we incrememt $u_f$ from the intersection point (i.e. move right along a fixed $u_{in}$ curve), the firing rate decreases, which decreases $u_f$, which cancels our initial increment.

## Steady state firing rate - numerical solution

We can do find the intersection between $f(u_f;u_{in})$ and $u_f=\alpha_ff$ numerically. To find the intersection, we use binary search (See [Appendix](#How-not-to-find-the-adaptive-LIF-neuron-firing-rate-numerically) for why) on $f$ given a fixed $u_{in}$ to find where $f(u_f;u_{in})$ is consistent with $u_f=\alpha_ff$.

```python
tau_m = .01
tref = .005
xt = 1.
af = 1e-2

th_max_f = 1./tref  # theoretical maximum firing rate
max_u = 5.  # maximum u_in to consider
max_f = th_lif_fi(max_u, tau_m, tref, xt)
f = np.linspace(5., max_f, 19)
u = th_lif_if(f, tau_m, tref, xt)  # linear in firing rate

num_af = num_alif_fi(u, tau_m, tref, xt, af)

cc = [(float(i)/len(u), 0, 0) for i in xrange(len(u))]
fig = figure(figsize=(16,6))

ax = fig.add_subplot(121)
_u = np.linspace(0, max(u), 100)
ax.plot(_u, th_lif_fi(_u, tau_m, tref, xt), 'k', linewidth=2, label='nonadaptive LIF ' + r'$(u_f=0)$')
ax.plot(_u, num_alif_fi(_u, tau_m, tref, xt, af=af), 'b', linewidth=2, label='adaptive LIF ' + r'$(\alpha_f=%.1e)$' % (af))
ax.plot(u, num_af, 'bo', mfc='none', ms=8)
ax.set_color_cycle(cc)
for idx, u_val in enumerate(u):
    f = th_lif_fi(u_val, tau_m, tref, xt)
    ax.plot([u_val, u[-1]], [f, f], c=cc[idx])
    ax.plot([u_val, u_val], [0, f], c=cc[idx], ls=':')
ax.set_ylim(0, th_max_f)
ax.set_xlim(0, max_u*1.001)
ax.legend(loc='upper left')
ax.set_xlabel(r'$u_{in}$', fontsize=20)
ax.set_ylabel(r'$f(u_{in})$ (spks / s)', fontsize=20)

f = np.linspace(0, th_max_f, 201)
uf = af*f

ax = fig.add_subplot(122, color_cycle=cc)
for idx, u_val in enumerate(u):
    net_u = u_val - uf
    f_ss = th_lif_fi(net_u, tau_m, tref, xt)
    line = ax.plot(uf, f_ss)[0]
    if u_val in (u[0], u[-1]):
        line.set_label(r'$u_{in}=%.2f$' % u_val)
iter_uf = af*num_af
ax.plot(iter_uf, num_af, 'bo', mfc='none', ms=8)
ax.plot(uf, f, 'b', lw=2, label=r'$u_f=\alpha_f f$')
ax.legend(loc='upper left', fontsize=16)
ax.set_xlabel(r'$u_f$', fontsize=20)
ax.set_ylabel(r'$f(u_{in},\ u_f)$ (spks / s)', fontsize=20)
fig.suptitle('Numerically solving for the adaptive LIF firing rate', fontsize=16);
```

The blue circles on the left plot correspond to the blue circles on the right plot. The numerical method finds the intersection of $f(u_f;u_{in})$ and $u_f=\alpha_ff$. Linear approximation is another way of estimating the intersection and is discussed in the [Appendix](#Steady-state-firing-rate---linear-approximation).

## Steady state firing rate validation

First, how long until the neuron has reached steady state?

```python
dt = .0001
T = 3.
tau_m = .01
tau_f = .01
tref = .005
xt = 1.
af = 1e-2

th_max_f = 1./tref  # theoretical maximum firing rate
max_u = 5.  # maximum u_in to consider
max_f = th_lif_fi(max_u, tau_m, tref, xt)
f = np.linspace(5., max_f, 5)
u = th_lif_if(f, tau_m, tref, xt)  # linear in firing rate

num_af = num_alif_fi(u, tau_m, tref, xt, af, tauf=tau_f, ref_fdecay=False)

obs_af = []
for u_val, num_af_val in zip(u, num_af):
    if num_af_val < .1:  # would require too long to simulate
        obs_af.append(0.)
        continue
    T_af = 1./num_af_val  # expected interspike interval
    run_time = 5.*tau_f+10.*T_af  # run long enough to reach steady state and collect a few spikes
    u_in = u_val+np.zeros(int(np.ceil(run_time/dt)))
    spike_times = run_alifsoma(dt, u_in, tau_m, tref, xt, af, tau_f)
    isi = np.diff(spike_times)
    if (isi[-2]-isi[-1])/isi[-2] > .01:
        print('Warning: greater than 1% change in isi between last two isi. ' +
              'Probably has not reached steady state for u_in=%.2f' % u_val)
    obs_af.append(1./isi[-1])

cc = [(float(i)/len(u), 0, 0) for i in xrange(len(u))]
fig = figure(figsize=(16,6))

ax = fig.add_subplot(121)
_u = np.linspace(0, max(u), 100)
lif_fi = th_lif_fi(_u, tau_m, tref, xt)
alif_fi = num_alif_fi(_u, tau_m, tref, xt, af=af, tauf=tau_f, ref_fdecay=False)
ax.plot(_u, lif_fi, 'k', linewidth=2, label='nonadaptive LIF ' + r'$(u_f=0)$')
ax.plot(_u, alif_fi, 'b', linewidth=2, label='adaptive LIF ' + r'$(\alpha_f=%.1e)$' % (af))
ax.plot(u, num_af, 'bo', mfc='none', ms=8)
ax.plot(u, obs_af, 'bx', mfc='none', ms=8)
ax.set_ylim(0, th_max_f)
ax.set_xlim(0, max_u*1.001)
ax.legend(loc='upper left')
ax.set_xlabel(r'$u_{in}$', fontsize=20)
ax.set_ylabel(r'$f(u_{in})$ (spks / s)', fontsize=20)
ax.set_title('Adaptive LIF firing rate\nnumerical solution vs simulation', fontsize=16);
```

## Steady state firing rate - linear approximation

We can use the linear approximation of the tuning curve equation from the LIF [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/neuron/LIF_neuron.ipynb#Linear-approximation-of-the-LIF-firing-rate-equation) to estimate the effect of adaptation on the tuning curve.

\begin{align}
f(u_{in}; a) &\approx \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}-\alpha_f f}\right)} \\
 &\approx k_0 + k_1(u_{in}-\alpha_ff) \\
\end{align}

where

\begin{align}
k_0 &= \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{a}\right)} - k_1a\\
k_1 &= \frac{\tau x_t}{\left(t_{ref}-\tau\ln\left(1-\frac{x_t}{a}\right)\right)^2a(a-x_t)} \\
\end{align}

$a$ is the region of $u_{in}-\alpha f$ that we are approximating around. We will set $a=u_{in}$. That is, we are linearly approximating from the nonadaptive ($u_f=0$) LIF firing rate equation.

\begin{align}
f &= k_0 + k_1(u_{in}-u_f) \\
 &= k_0 + k_1u_{in}-k_1u_f & \text{linearly approximate $f(u_f; u_{in})$}\\
 &= k_0 + k_1u_{in}-k_1\alpha_ff & u_f=\alpha f\\
(1+k_1\alpha_f)f &= k_0 +k_1u_{in} \\
f &= \frac{k_0 +k_1u_{in}}{1+k_1\alpha_f} & \text{intersection of $f(u_f; u_{in})$ and $u_f=\alpha f$}\\
\end{align}


```python
def taylor1_lif_k0_k1(a, tau, tref, xt):
    k1 = tau*xt/((tref-tau*np.log(1-xt/a))**2*a*(a-xt))
    k0 = th_lif_fi(a, tau, tref, xt) - k1*a
    return k0, k1

tau_m = .01
tref = .005
xt = 1.
af = 1e-2

th_max_f = 1./tref  # theoretical maximum firing rate
max_u = 5.  # maximum u_in to consider
max_f = th_lif_fi(max_u, tau_m, tref, xt)
f = np.linspace(5., max_f, 15)
u_in = th_lif_if(f, tau_m, tref, xt)  # linear in firing rate

uf_f = np.linspace(0, th_max_f, 201)
uf = af*uf_f

k0, k1 = taylor1_lif_k0_k1(u_in, tau_m, tref, xt)
taylor1_f_uf = np.zeros((u_in.shape[0], uf.shape[0]))
for idx, u_val in enumerate(u_in):
    taylor1_f_uf[idx, :] = k0[idx]+k1[idx]*u_val-k1[idx]*uf
taylor1_af = (k0+k1*u_in)/(1+k1*af)

cc = [(float(i)/len(u_in), 0, 0) for i in xrange(len(u_in))]
fig = figure(figsize=(16,6))

ax = fig.add_subplot(121)
_u = np.linspace(0, max(u_in), 100)
idx = _u>xt
k0, k1 = taylor1_lif_k0_k1(_u[idx], tau_m, tref, xt)
_taylor1_af = (k0+k1*_u[idx])/(1+k1*af)

ax.plot(_u, th_lif_fi(_u, tau_m, tref, xt), 'k', linewidth=2, label='nonadaptive LIF ' + r'$(u_f=0)$')
ax.plot(_u[idx], _taylor1_af, 'b', linewidth=2, label='adaptive LIF ' + r'$(\alpha_f=%.1e)$' % (af))
ax.plot(u_in, taylor1_af, 'bo', mfc='none', ms=8)
ax.set_color_cycle(cc)
for idx, u_val in enumerate(u_in):
    f = th_lif_fi(u_val, tau_m, tref, xt)
    ax.plot([u_val, u_in[-1]], [f, f], c=cc[idx])
    ax.plot([u_val, u_val], [0, f], c=cc[idx], ls=':')
ax.set_ylim(0, th_max_f)
ax.set_xlim(0, max_u*1.001)
ax.legend(loc='upper left')
ax.set_xlabel(r'$u_{in}$', fontsize=20)
ax.set_ylabel(r'$f(u_{in})$ (spks / s)', fontsize=20)

ax = fig.add_subplot(122)
for idx, u_val in enumerate(u_in):
    net_u = u_val - uf
    f_ss = th_lif_fi(net_u, tau_m, tref, xt)
    line = ax.plot(uf, f_ss, color=cc[idx])[0]
    if u_val in (u_in[0], u_in[-1]):
        line.set_label(r'$u_{in}=%.2f$' % u_val)
    ax.plot(uf, taylor1_f_uf[idx, :], color=cc[idx], ls=':')
iter_uf = af*taylor1_af
ax.plot(iter_uf, taylor1_af, 'bo', mfc='none', ms=8)
ax.plot(uf, uf_f, 'b', lw=2, label=r'$u_f=\alpha_f f$')
ax.legend(loc='upper left', fontsize=16)
ax.set_ylim(0, th_max_f)
ax.set_xlabel(r'$u_f$', fontsize=20)
ax.set_ylabel(r'$f(u_f;\ u_{in})$ (spks / s)', fontsize=20)
fig.suptitle('Linearly approximating the adaptive LIF firing rate', fontsize=16);
```

On the right plot, we see that linear approximation consistently overestimates the intersection of $f(u_f;u_{in})$ and $u_f=\alpha_ff$, but the approximation gets better for higher firing rates.
