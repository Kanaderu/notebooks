# Background

The LIF neuron firing rate is described by

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\ln\left(1-\frac{x_t}{u_{net}}\right)} \\
\end{align}

where
 - $t_{ref}$ is the refractory period
 - $\tau_m$ is the soma membrane time constant
 - $x_t$ is the threshold (typically set to $1$)
 - $u_{net}$ is the net input ($u_{in}-u_f$) to the soma
 
As its name implies, the rate-based adaptive LIF neuron does not output spikes. Instead, for a given net input it outputs a real-valued rate. This rate is the rate at which a spiking LIF neuron would spike given the same net input. 

![rate-based adaptive LIF neuron](figures/rate_adaptive_LIF.png "rate-based adaptive LIF neuron")

For the nonadaptive LIF neuron, the net input to the soma is simply the input to the neuron. That is, $u_{net}=u_{in}$. However, for the adaptive LIF neuron, there is also an adaptive negative feedback $u_f$ input, so

\begin{align}
u_{net} &= u_{in}-u_f \\
\end{align}

The adaptive feedback term has the equivalent dynamics of a first order low-pass filter (i.e. a simple synapse) so that

\begin{align}
\tau_f\dot{u}_f &= -u_f+\alpha_f f \\
\end{align}

Since $u_{net} = u_{in} - u_f$ and if we assume $u_{in}$ is constant,

\begin{align}
\dot{u}_{net} &= -\dot{u}_f \\
\end{align}

Therefore,

\begin{align}
\tau_f\dot{u}_{net} &= u_f-\alpha_f f \\
 &= -u_{net}+u_{in}-\alpha_f f \\
\end{align}

# Analysis

## Steady state

At steady state, $\dot{u}_{net}=0$ so

\begin{align}
u_f &= \alpha_ff \\
\end{align}

and

\begin{align}
u_{net} &= u_{in} - \alpha_ff \\
\end{align}

The steady state firing rate would then be

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\ln\left(1-\frac{x_t}{u_{in}-\alpha_f f}\right)} \\
\end{align}

There isn't an analytical solution for $f$ for this equation. However we can find the firing rate numerically using the bisection method in a similar manner to the mean feedback approximation of the spiking adaptive LIF [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/neuron/adaptive_LIF_neuron.ipynb#Mean-feedback-approximation---numerical-solution) (without compensating for the refractory period decay of the feedback in $\alpha_f$). However, we will see that this solution isn't unique when analyzing the dynamics.

## Dynamics

We would like to know $\frac{df}{dt}$. Our approach is to find $\frac{\partial f}{\partial u_{net}}$ and $\frac{\partial u_{net}}{\partial t}$ and then apply the chain rule, which states that

$$\frac{df}{dt}=\frac{\partial f}{\partial u_{net}}\frac{\partial u_{net}}{\partial t}$$

For $\frac{df}{dt}$, from

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\ln\left(1-\frac{x_t}{u_{net}}\right)} \\
\end{align}

we find

\begin{align}
\frac{\partial f}{\partial u_{net}} &= 
    \frac{-1}{\left(t_{ref}-\tau_m\ln\left(1-\frac{x_t}{u_{net}}\right)\right)^2}
    \left(\frac{-\tau_m}{1-\frac{x_t}{u_{net}}}\right)
    \left(\frac{x_t}{{u_{net}}^2}\right) \\
 &= f^2\frac{\tau_m}{\frac{u_{net}-x_t}{u_{net}}}\frac{x_t}{{u_{net}}^2} \\
 &= f^2\frac{\tau_mx_t}{u_{net}(u_{net}-x_t)} \\
\end{align}

For $\frac{\partial u_{net}}{\partial t}$, recall that $u_{net} = u_{in} - u_f$, so

\begin{align}
\dot{u}_{net} &= -\dot{u}_f \\
\end{align}

where

\begin{align}
\tau_f\dot{u}_f &= -u_f+\alpha_f f \\
\end{align}

Therefore,

\begin{align}
\tau_f\dot{u}_{net} &= u_f-\alpha_f f \\
 &= -u_{net}+u_{in}-\alpha_f f \\
\end{align}

Using $\frac{df}{dt}=\frac{\partial f}{\partial u_{net}}\frac{\partial u_{net}}{\partial t}$,

\begin{align}
\frac{df}{dt} &= 
    f^2\frac{\tau_mx_t}{u_{net}(u_{net}-x_t)}
    \frac{1}{\tau_f}(u_f-\alpha_f f) \\
\tau\frac{df}{dt} &= f^2(u_f-\alpha_f f) \\
 &= u_ff^2 - \alpha_ff^3  \\
\end{align}

where $\tau = \frac{\tau_f}{\tau_mx_t}u_{net}(u_{net}-x_t)$

<div style="display:block; border: 1px solid black; padding: 15px;background-color: #EEEEFF;margin: 10px">
To summarize the dynamics,

\begin{align}
\tau_f\frac{du_{net}}{dt} &= -u_{net}+u_{in}-\alpha_f f \\
\end{align}

and

\begin{align}
\tau\frac{df}{dt} &= u_ff^2 - \alpha_ff^3 \\
 \end{align}

where $\tau = \frac{\tau_f}{\tau_mx_t}u_{net}(u_{net}-x_t)$
</div>

At this point, we can see that the steady state value firing rate found above is not unique across $u_{net}$ and $f$. First, considering $\frac{du_{net}}{dt}=0$,

\begin{align}
0 &= u_f - \alpha_ff \\
u_f &= \alpha_ff \\
\end{align}

Now considering $\frac{df}{dt}=0$, 

\begin{align}
0 &= u_ff^2 - \alpha_ff^3 \\
 &= f^2(u_f - \alpha_ff) \\
\end{align}

which has solutions at 

\begin{align}
f&=0, & f&=\frac{u_f}{\alpha_f} \\
\end{align}

which is reassuringly consistent with our solution for $\frac{du_{net}}{dt}=0$ and also implies that there are infinite, linearly related $(u_{net}, f)$ steady state pairs. 

Here's an interactive phase plot you can play with.

```python
def interactive_phase_u_f(u_in=3.5, logaf=-2.):
    af = 10**logaf
    fig, ax = phase_u_f(tau_m, tref, xt, af, tau_f, u_in=u_in)
    f_open = th_lif_fi(u_in, tau_m, tref, xt)
    ax.plot(u_in, f_open, 'ro')
    ax.plot([u_in, u_in], [0, f_open], 'r:')
    f_closed = num_ralif_fi(u_in, tau_m, tref, xt, af)
    gain = f_open/f_closed
    ax.set_title('gain = open loop rate / closed loop rate = %.3f' % gain)

display(interactive(interactive_phase_u_f, u_in=(xt*1.01, 10.), logaf=(-6., 1.)))
```

We define the gain

\begin{align}
A &= \frac{f_{open}}{f_{closed}}
\end{align}

where $f_{open}$ and $f_{closed}$ are the open (without adaptive feedback) and closed loop (with adaptive feedback) rates, respectively.

See that if we initialize the rate-based adaptive LIF neuron to the open loop firing rate, it will follow the open loop tuning curve until intersecting with the $f=-\alpha_f(u_{in}-u_{net})$ line.

# Rate vs spiking implementation

The rate-based adaptive LIF neuron is useful because we can analyze its dynamics. 

How does the rate neuron compare to the spiking neuron? We'll feed both into synapses and compare the outputs of the synapses.

```python
tau_m = .005
tref = .001
tau_f = .1
tau_syn = tau_f

dt = 1e-3
T = 5*tau_f

rate_v_spiking(3.5, tau_m, tref, xt, af, tau_f, tau_syn/20., dt, T);
```

Here we see how the adaptive neuron initially spikes quickly but then settles down to its steady state firing rate. More importantly, see how the synapse dynamics resulting from the spiking adaptive neuron match the dynamics resulting from the rate-based adaptive neuron. Note that I am matching the steady-state outputs of the different neuron types by adjusting the inputs to the rate-based adaptive LIF neuron and the standard LIF neuron.

Matching the feedback time constant and the synaptic time constant, we see how the adaptive neuron always drives the synapse faster. Here is an interactive plot.

```python
def interactive_rate_v_spiking(alif_u_in=3.5):
    rate_v_spiking(alif_u_in, tau_m, tref, xt, af, tau_f, tau_syn, dt, T, normalize=True)
display(interactive(interactive_rate_v_spiking, alif_u_in=(xt*1.01, 10.)))
```

```python
u_ins = [1.2, 2.0, 3.0,]

for u_in in u_ins:
    rate_v_spiking(u_in, tau_m, tref, xt, af, tau_f, tau_syn, dt, T)
```

See how the speedup actually decreases with as input increases. This is easier to see with the interactive plot, but on the static plots, see how the black line gets closer to the blue trace with higher inputs. Remember that we're looking at the output of a synapse. The benefit of the rate-based adaptive LIF neuron is that we can see the instantaneous rate. Here is an interactive plot of the instantaneous rate over time.

```python
from IPython.html.widgets import RadioButtonsWidget

def interactive_f_traj(u_in=3.5, log10dt=-3, normalizey=True):
    dt = 10**log10dt
    f_traj(u_in, tau_m, tref, xt, af, tau_f, dt, normalizey=normalizey)
display(interactive(
    interactive_f_traj, u_in=(1.01*xt, 5.), log10dt=(-5,-3),
    normalizey=RadioButtonsWidget(values=[True, False], value=True)))
```
