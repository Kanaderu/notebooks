# Notebook description

This notebook examines the rate-based adaptive LIF neuron. Specifically
 - What is the steady-state rate of a rate-based adaptive neuron?
 - What are the dynamics of the neuron's output rate?

# Analysis

The LIF neuron firing rate is described by

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\left(1-\frac{1}{u}\right)} \\
\end{align}

where
 - $t_{ref}$ is the refractory period
 - $\tau_m$ is the soma membrane time constant
 - $u$ is the total input to the soma

Without adaptation, $u=u_{in}$. However, with adaptation,

\begin{align}
u &= u_{in}-u_f \\
\end{align}

where $u_f$ is the adaptive feedback, and

\begin{align}
\tau_f\dot{u}_f &= -u_f+\alpha_f f \\
\end{align}

To answer what is the steady state firing rate, we look at the dynamics of $u$. If we assume $u_{in}$ is held constant,

\begin{align}
\dot{u} &= -\dot{u}_f \\
 &= -\frac{1}{\tau_f}u_f + \frac{\alpha_f}{\tau_f} f \\
\end{align}

At steady state, $\dot{u}=0$ so

\begin{align}
\alpha_ff &= u_f \\
\end{align}

Substituting the steady state feedback into the firing rate equation,

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\ln\left(1-\frac{1}{u_{in}-\alpha_f f}\right)} \\
\end{align}

Solving for $f$ would yield the steady-state firing rate. Unfortunately, there isn't an analytical solution to this equation.

```python
tau_m=.1
tref=.002
xt=1.
a_f=1.
tau_f=.1
n_f=15
max_u=200.
dt=1e-3
inc = -np.expm1(-dt/tau_f)
decay = np.expm1(-dt/tau_f)+1.
alpha=.7

u = np.sort(np.linspace(0, max_u, 100).tolist() + [xt, 1.001*xt])
f = th_lif_fi(u_fine, tau_m, tref, xt)
max_f = max(f)
min_f = max(f)/10.

fig = figure(figsize=(16,6))
ax = fig.add_subplot(121)
ax.plot(u, f, 'k', lw=1)

f, f_step = np.linspace(min_f, max_f, n_f, retstep=True)

U_in, F, dU, dF = [], [], [], []
for f_open in f:
    f_arr = np.arange(f_open, 0., -f_step)
    u_in_arr = th_lif_if(f_open, tau_m, tref, xt)*np.ones(f_arr.shape)
    u_arr = th_lif_if(f_arr, tau_m, tref, xt)
    u_f_arr = u_in_arr - u_arr
    du_f_arr = -inc*u_f_arr + inc*a_f*f_arr
    df_arr = th_lif_fi(u_in_arr-(u_f_arr+du_f_arr), tau_m, tref, xt)-f_arr
    
    U_in += u_in_arr.tolist()
    F += f_arr.tolist()
    dU += (-du_f_arr).tolist()
    dF += df_arr.tolist()

bcmap = make_blue_cmap(low=1., high=0.)
ax.quiver(U_in, F, dU, dF, dF, cmap=bcmap, alpha=alpha, angles='xy', pivot='middle')
ax.set_xlabel(r'$u_{in}\ ;\ u$', fontsize=20)
ax.set_ylabel(r'$f$', fontsize=20)


u_in = th_lif_if(f, tau_m, tref, xt)
rcmap = make_red_cmap()
cc = make_color_cycle(range(n_f), rcmap)
for idx, (u_val, f_val) in enumerate(zip(u_in, f)):
    ax.plot([u_val, u_in[-1]], [f_val, f_val], c=cc[idx])
    ax.plot([u_val, u_val], [0, f_val], c=cc[idx], ls=':')

U_f, F, dU_f, dF = [], [], [], []

n_uf_fine = 30
u_f_fine = np.zeros((n_uf_fine, n_f))
for idx, u_val in enumerate(u_in):
    u_f_fine[:, idx] = np.linspace(0, u_val-xt, n_uf_fine)
f_fine = th_lif_fi(u_in-u_f_fine, tau_m, tref, xt)
ax = fig.add_subplot(122, sharey=ax)
ax.set_color_cycle(cc)
ax.plot(u_f_fine, f_fine)

for idx, u_val in enumerate(u_in):
    f_open = th_lif_fi(u_val, tau_m, tref, xt)
    f_arr = np.arange(f_open, 0., -f_step)
    u_net = th_lif_if(f_arr, tau_m, tref, xt)
    u_f_arr = u_val - u_net  
    du_f_arr = -inc*u_f_arr + a_f*inc*f_arr
    df_arr = th_lif_fi(u_val-(u_f_arr+du_f_arr), tau_m, tref, xt)-f_arr
    
    U_f += u_f_arr.tolist()
    F += f_arr.tolist()
    dU_f += du_f_arr.tolist()
    dF += df_arr.tolist()
ax.quiver(U_f, F, dU_f, dF, dF, cmap=bcmap, alpha=alpha, angles='xy', pivot='middle')

u_f = np.array([0, max_u])
f = u_f / a_f
ax.plot(u_f, f, 'b')
ax.set_xlabel(r'$u_f$', fontsize=20);
```
