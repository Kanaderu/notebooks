# Background

As its name implies, the rate-based adaptive LIF neuron does not output spikes. Instead, for a given net input (including feedback) it outputs a real-valued rate. This rate is the rate at which a spiking LIF neuron would spike given the same net input. 

The LIF neuron firing rate is described by

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\left(1-\frac{x_t}{u_{net}}\right)} \\
\end{align}

where
 - $t_{ref}$ is the refractory period
 - $\tau_m$ is the soma membrane time constant
 - $x_t$ is the threshold (typically set to $1$)
 - $u_{net}$ is the net input to the soma

For the nonadaptive LIF neuron, the net input to the soma is simply the input to the neuron. That is, $u_{net}=u_{in}$. However, for the adaptive LIF neuron, there is also an adaptive negative feedback $u_f$ input, so

\begin{align}
u_{net} &= u_{in}-u_f \\
\end{align}

The adaptive feedback term has the equivalent dynamics of a first order low-pass filter (i.e. a simple synapse) so that

\begin{align}
\tau_f\dot{u}_f &= -u_f+\alpha_f f \\
\end{align}

# Analysis

## Steady state

To answer what is the steady state firing rate, we look at the dynamics of $u$. If we assume $u_{in}$ is held constant,

\begin{align}
\dot{u}_{net} &= -\dot{u}_f \\
 &= -\frac{1}{\tau_f}u_f + \frac{\alpha_f}{\tau_f} f \\
\end{align}

At steady state, $\dot{u}_{net}=0$ so

\begin{align}
u_f &= \alpha_ff \\
\end{align}

and

\begin{align}
u_{net} &= u_{in} - \alpha_ff \\
\end{align}

The steady state firing rate would then be

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\ln\left(1-\frac{x_t}{u_{in}-\alpha_f f}\right)} \\
\end{align}

Unfortunately, there isn't an analytical solution for $f$ for this equation. However we can find the firing rate numerically using the bisection method in a similar manner to the mean feedback approximation of the spiking adaptive LIF [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/neuron/adaptive_LIF_neuron.ipynb#Mean-feedback-approximation---numerical-solution). In contrast to the mean feedback approximation method, we don't need to compensate for the refractory period decay of the feedback in $\alpha_f$.

## Dynamics

We would like to know $\frac{df}{dt}$. Our approach is to find $\frac{\partial f}{\partial u_{net}}$ and $\frac{\partial u_{net}}{\partial t}$ and then apply the chain rule, which states that

$$\frac{df}{dt}=\frac{\partial f}{\partial u_{net}}\frac{\partial u_{net}}{\partial t}$$

For $\frac{df}{dt}$, from

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\ln\left(1-\frac{x_t}{u_{net}}\right)} \\
\end{align}

we find

\begin{align}
\frac{\partial f}{\partial u_{net}} &= 
    \frac{-1}{\left(t_{ref}-\tau_m\ln\left(1-\frac{x_t}{u_{net}}\right)\right)^2}
    \left(\frac{-\tau_m}{1-\frac{x_t}{u_{net}}}\right)
    \left(\frac{x_t}{{u_{net}}^2}\right) \\
 &= f^2\frac{\tau_m}{\frac{u_{net}-x_t}{u_{net}}}\frac{x_t}{{u_{net}}^2} \\
 &= f^2\frac{\tau_mx_t}{u_{net}(u_{net}-x_t)} \\
\end{align}

For $\frac{\partial u_{net}}{\partial t}$, recall that $u_{net} = u_{in} - u_f$, so

\begin{align}
\dot{u}_{net} &= -\dot{u}_f \\
\end{align}

where

\begin{align}
\tau_f\dot{u}_f &= -u_f+\alpha_f f \\
\end{align}

Therefore,

\begin{align}
\tau_f\dot{u}_{net} &= u_f-\alpha_f f \\
 &= -u_{net}+u_{in}-\alpha_f f \\
\end{align}

Using $\frac{df}{dt}=\frac{\partial f}{\partial u_{net}}\frac{\partial u_{net}}{\partial t}$,

\begin{align}
\frac{df}{dt} &= 
    f^2\frac{\tau_mx_t}{u_{net}(u_{net}-x_t)}
    \frac{1}{\tau_f}(-u_{net}+u_{in}-\alpha_f f) \\
 &= f^2\frac{\tau_mx_t}{\tau_f}\frac{1}{u_{net}(u_{net}-x_t)}
    (-u_{net}+u_{in}-\alpha_f f) \\
 &= k_2f^2 + k_3f^3  \\
\end{align}

where 

\begin{align}
k_2 &= \frac{\tau_mx_t}{\tau_f}\frac{(u_{in}-u_{net})}{u_{net}(u_{net}-x_t)} \\
k_3 &= \frac{\tau_mx_t}{\tau_f}\frac{-\alpha_f}{u_{net}(u_{net}-x_t)} \\
\end{align}
