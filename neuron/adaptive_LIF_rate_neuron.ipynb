# Background

As its name implies, the rate-based adaptive LIF neuron does not output spikes. Instead, for a given net input (including feedback) it outputs a real-valued rate. This rate is the rate at which a spiking LIF neuron would spike given the same net input. 

The LIF neuron firing rate is described by

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\left(1-\frac{1}{u}\right)} \\
\end{align}

where
 - $t_{ref}$ is the refractory period
 - $\tau_m$ is the soma membrane time constant
 - $u$ is the net input to the soma

For the nonadaptive LIF neuron, the net input to the soma is simply the input to the neuron. That is, $u=u_{in}$. However, for the adaptive LIF neuron, there is also an adaptive negative feedback $u_f$ input, so

\begin{align}
u &= u_{in}-u_f \\
\end{align}

The adaptive feedback term has the equivalent dynamics of a first order low-pass filter (i.e. a simple synapse) so that

\begin{align}
\tau_f\dot{u}_f &= -u_f+\alpha_f f \\
\end{align}

# Analysis

To answer what is the steady state firing rate, we look at the dynamics of $u$. If we assume $u_{in}$ is held constant,

\begin{align}
\dot{u} &= -\dot{u}_f \\
 &= -\frac{1}{\tau_f}u_f + \frac{\alpha_f}{\tau_f} f \\
\end{align}

At steady state, $\dot{u}=0$ so

\begin{align}
\alpha_ff &= u_f \\
\end{align}

Substituting the steady state feedback into the firing rate equation,

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\ln\left(1-\frac{1}{u_{in}-\alpha_f f}\right)} \\
\end{align}

Solving for $f$ would yield the steady-state firing rate. Unfortunately, there isn't an analytical solution to this equation. However we can find the firing rate numerically using the bisection method as in the adaptive LIF notebook.
