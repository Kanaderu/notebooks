# Notebook description

This notebook examines the rate-based adaptive LIF neuron. Specifically
 - What is the steady-state rate of a rate-based adaptive neuron?
 - What are the dynamics of the neuron's output rate?

# Analysis

The LIF neuron firing rate is described by

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\left(1-\frac{1}{u}\right)} \\
\end{align}

where
 - $t_{ref}$ is the refractory period
 - $\tau_m$ is the soma membrane time constant
 - $u$ is the total input to the soma

Without adaptation, $u=u_{in}$. However, with adaptation,

\begin{align}
u &= u_{in}-u_f \\
\end{align}

where $u_f$ is the adaptive feedback, and

\begin{align}
\tau_f\dot{u}_f &= -u_f+\alpha_f f \\
\end{align}

To answer what is the steady state firing rate, we look at the dynamics of $u$. If we assume $u_{in}$ is held constant,

\begin{align}
\dot{u} &= -\dot{u}_f \\
 &= -\frac{1}{\tau_f}u_f + \frac{\alpha_f}{\tau_f} f \\
\end{align}

At steady state, $\dot{u}=0$ so

\begin{align}
\alpha_ff &= u_f \\
\end{align}

Substituting the steady state feedback into the firing rate equation,

\begin{align}
f &= \frac{1}{t_{ref}-\tau_m\ln\left(1-\frac{1}{u_{in}-\alpha_f f}\right)} \\
\end{align}

Solving for $f$ would yield the steady-state firing rate. Unfortunately, there isn't an analytical solution to this equation.

```python
tau_m = .05
tref = .002
xt = 1.
a_f = 1.
tau_f = .01
dt = .001

inc = -np.expm1(-dt/tau_f)
decay = np.expm1(-dt/tau_f)+1.

n_u = 20
n_f = 10

max_u = 100.

u = np.linspace(0, max_u, n_u)
f = th_lif_fi(u, tau_m, tref, xt)

U_in = np.tile(u, (n_f, 1))
U = np.zeros_like(U_in)  # total input
F = np.zeros_like(U_in)
dU = np.zeros_like(U_in)
dF = np.zeros_like(U_in)

for idx, u_val in enumerate(u):
    f_open = th_lif_fi(u_val, tau_m, tref, xt)
    F[:, idx] = np.linspace(0, f_open, n_f)
    f_idx = F[:, idx] > 0.
    U[f_idx, idx] = th_lif_if(F[f_idx, idx], tau_m, tref, xt)  # net input to required
    u_adapt = U[f_idx, idx] - u_val
    adapt = -a_f*F[f_idx,idx]
    dU[f_idx, idx] = inc*(adapt - u_adapt)
    dF[f_idx, idx] = th_lif_fi(U[f_idx, idx]+dU[f_idx, idx], tau_m, tref, xt) - F[f_idx, idx] 

C = dF
cmap = make_red_cmap(low=1., high=0.)
fig = figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(u, f, 'k', lw=2)
ax.quiver(U_in, F, dU, dF, C, cmap=cmap)
ax.set_xlim(0, max_u)
ax.set_xlabel(r'$u$', fontsize=20)
ax.set_ylabel(r'$f$', fontsize=20);
```
