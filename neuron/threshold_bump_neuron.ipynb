# Notebook description

This notebook explores an alternative to the LIF neuron.  Instead of computing the dynamics of the LIF neuron's soma, we could opt for the soma to simply be a function that outputs spikes with minimal dynamics. A functional spiking instead of dynamical spiking implementation could make it easier to implement a neuron in digital design.

# Threshold neuron description

The threshold neuron is composed of a standard exponential synapse and a simple thresholding soma. The thresholding soma has 3 parameters:
 - a lower threshold
 - an upper threshold
 - an interspike interval (in units of dt)

When the input to the soma is constant and between the lower and upper thresholds, the soma spikes uniformly at a rate given by the interspike interval. The simplicity makes it very easy to design and validate. In addition, when the exponential synapse is coupled to the soma, the synapse provides noisy input to the soma, effectively smoothing off the neuron's tuning curve. If one knows the distribution of the synapse's value given the input spike train statistics, it shouldn't be too hard to find the distribution of the soma's spiking rate, which would allow us to find and use the noisy tuning curve without collecting any data in training decoding weights.

# Producing a multi-dimensional bump tuning curve

Bump shaped tuning curves make it easier to decode many functions. However, given the standard method for calculating the input current to a neuron, it is impossible to get a truly multi-dimensional, bump-shaped tuning curve in more than 1-dimension.

The standard way of calculating the input current is 

$$
J = \mathbf{e}\cdot\mathbf{x} + x_0,
$$

where $J$ is the input current, $\mathbf{e}$ is the neuron's encoding vector, $\mathbf{x}$ is the input, and $x_0$ is the bias current. From here we can see that any change in $\mathbf{x}$ that is orthoganol to $\mathbf{e}$ will not affect the input current to the neuron, and so it is impossible for the neuron to be bump-shaped along the direction orthoganol to $\mathbf{e}$.   

If we modify the encoding scheme, we can use the threshold neuron (or any neuron with a bump-like tuning curve) to produce a multi-dimensional bump-shaped tuning curve. By considering the difference between $\mathbf{x}$ and $\mathbf{e}$, we can produce a bump in the input current. That is, we can use

$$
J = \|\mathbf{e}-\mathbf{x}\|_i,
$$

where $i$ indicates which norm to use. The L1 norm is the sum of absolute values and should be easy to implement in hardware. Alex said the L2 should also be feasible. With this encoding scheme and a neuron that has a bump tuning curve, we can create a tuning curve that is a bump localized in all dimensions, not just along the encoding vector's direction.

# Simulations

I show the following with simulations:
 - An exponential synapse is noisy
 - The threshold neuron has a box-shaped tuning curve given direct current injection.
 - The threshold neuron has a bump-shaped tuning curve given noisy spiking input.
 - The threshold neuron supports a saturating tuning curve that looks like a sigmoid.
 - With the vector difference encoding scheme and the threshold neuron, we can construct multi-dimensional bump tuning curves.

```python
%pylab inline
import numpy as np
from numpy.linalg import norm
from numpy.random import poisson
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
```

```python
# define models to simulate
class Synapse(object):
    def __init__(self, dt=.001, tau=.01, w=1.):
        self.dt = dt
        self.tau = tau
        self.w = w
        
        self.state = 0.
        self.a = np.exp(-self.dt/self.tau) # decay
        self.u = 1./self.dt # input unit
        
    def reset(self):
        self.state = 0.
        
    def step(self):
        self.state = self.a*self.state
        
    def spike_in(self, nspikes=1):
        self.state += nspikes*self.w*(1.-self.a)*self.u
        

class ThresholdNeuron(object):
    """Threshold neuron fires uniformly when its synaptic input is within a specified range.
    
    Range specified by t_low and t_high
    
    Parameters
    ----------
    dt : float
        timestep of neuron
    t_low : float or None
        lower threshold of neuron
    t_high : float or None
        upper threshold of neuron
    period: int
        period of uniform firing rate in units of dt
    """
    def __init__(self, dt, t_low, t_high, period, tau_syn=0.01, w_syn=1.):
        if t_low and t_high:
            assert t_low < t_high, "lower threshold must be less than upper threshold"
        assert isinstance(period, int), "period must be in units of dt"
        assert period > 0

        self.dt = dt
        self.t_low = t_low
        self.t_high = t_high
        self.T = period
        self.syn = Synapse(dt, tau_syn, w_syn)
        self.ctr = self.T
    
    def reset(self):
        self.ctr = self.T
        self.syn.reset()
    
    def _in_range(self, in_cur):
        in_range = True
        if self.t_low and in_cur < self.t_low:
            in_range = False
        if self.t_high and in_cur > self.t_high:
            in_range = False
        return in_range
        
    def step(self, spikes_in=0, direct_in=0.):
        self.syn.step()
        self.syn.spike_in(spikes_in)
        in_cur = direct_in + self.syn.state

        if (self._in_range(in_cur)):
            self.ctr -= 1
            if self.ctr == 0:
                self.ctr = self.T
                return 1
            
        return 0
    
class BumpThresholdNeuron(ThresholdNeuron):
    """BumpNeuron responds to multidimensional inputs"""
    def __init__(self, e_vector, dt, t_low, t_high, period, tau_syn=.01, w_syn=.001, norm=1):
        ThresholdNeuron.__init__(self, dt, t_low, t_high, period, tau_syn, w_syn)
        self.e = e_vector
        self.ndim = self.e.shape[0]
        self.syn = [Synapse(dt, tau_syn, w_syn) for i in xrange(self.ndim)]
        self.norm = norm
    
    def reset(self):
        self.ctr = self.T
        for dim in xrange(self.ndim):
            self.syn[dim].reset()

    def _update_syn(self, spikes_in):
        syn_in = np.zeros(self.ndim)
        for dim, syn in enumerate(self.syn):
            syn.step()
            syn.spike_in(spikes_in[dim])
            syn_in[dim] = syn.state
        in_cur_syn = norm(self.e-syn_in, ord=self.norm)
        return in_cur_syn
    
    def step(self, spikes_in=None, direct_in=None):
        in_cur = 0.
        if spikes_in is not None:
            in_cur += self._update_syn(spikes_in)
        in_cur_spikes = in_cur
        
        if direct_in is not None:
            in_cur += norm(self.e-direct_in, ord=self.norm)
        in_cur_direct = in_cur - in_cur_spikes
        
        if (self._in_range(in_cur)):
            self.ctr -= 1
            if self.ctr == 0:
                self.ctr = self.T
                return 1
            
        return 0
```

```python
# utilities for collecting data
def get_tuning(neuron, input_rates, n_steps):
    dt = neuron.dt
    n_inputs = len(input_rates)
    spikes = np.zeros((n_steps, n_inputs))
    for f_idx, f_in in enumerate(input_rates):
        lam = f_in * dt
        neuron.reset()
        for t_idx in xrange(n_steps):
            spks = poisson(lam)
            spiked = neuron.step(spikes_in=spks)
            if spiked:
                spikes[t_idx, f_idx] = 1

    spike_rates = np.sum(spikes, axis=0) / (float(n_steps)*dt)
    return spike_rates


def get_tuning_multid(neuron, n_steps, input_rates=None, direct_inputs=None):
    dt = neuron.dt
    if input_rates is not None:
        n_inputs = input_rates.shape[0]
        n_dim = input_rates.shape[1]
    elif direct_inputs is not None:
        n_inputs = direct_inputs.shape[0]
        n_dim = direct_inputs.shape[1]
    
    in_cur_direct = None
    spks = None
    spikes = np.zeros((n_steps, n_inputs))
    for in_idx in xrange(n_inputs):
        if input_rates is not None:
            lam = np.array([input_rates[in_idx][dim_idx] * dt for dim_idx in xrange(n_dim)])
        if direct_inputs is not None:
            in_cur_direct = direct_inputs[in_idx]
        neuron.reset()
        syn_state = np.zeros((n_steps, n_dim))
        for t_idx in xrange(n_steps):
            if input_rates is not None:
                spks = np.array([poisson(lam[dim_idx]) for dim_idx in xrange(n_dim)])
            spiked = neuron.step(spikes_in=spks, direct_in=in_cur_direct)
            if spiked:
                spikes[t_idx, in_idx] = 1
            for dim_idx in xrange(n_dim):
                syn_state[t_idx] = neuron.syn[dim_idx].state

    spike_rates = np.sum(spikes, axis=0) / (float(n_steps)*dt)
    return spike_rates

def unroll_inputs(in_x, in_y):
    """unrolls input grids. 2d support only"""
    assert in_x.shape == in_y.shape
    n_inputs_x = in_x.shape[0]
    n_inputs_y = in_x.shape[1]
    n_inputs = n_inputs_x * n_inputs_y
    in_unrolled = np.zeros((n_inputs, 2))
    idx = 0
    for x in xrange(n_inputs_x):
        for y in xrange(n_inputs_y): 
            in_unrolled[idx, 0] = in_x[x,y]
            in_unrolled[idx, 1] = in_y[x,y]
            idx += 1
    return in_unrolled
```

```python
# validate synapse
dt = .001
N = 10000 # time steps to simulate
f = 1000. # input spike rate

tau = .01
w = 1.
syn = Synapse(dt, tau, w)

lam = dt*f
syn_states = np.zeros(N)
for idx in xrange(N):
    syn.step()
    spks = poisson(lam)
    syn.spike_in(spks)
    syn_states[idx] = syn.state

t = dt*np.arange(N)
syn_ss = syn_states[t>5.*tau]
ss_mean = np.mean(syn_ss)
ss_std = np.std(syn_ss)

figure(figsize=(8,6))
plot(t, syn_states, '-')
axhline(ss_mean, color='r', linewidth=2, label='steady state mean')
axhline(f, color='g', linewidth=2, linestyle=':', label='input spike rate')
legend(loc='lower left')
xlabel('time (s)', fontsize=18)
ylabel('synapse state', fontsize=18)
title('The exponential synapse is noisy', fontsize=18);

figure(figsize=(8,6))
hist(syn_ss, bins=100, normed=True)
axvline(ss_mean, color='r', linewidth=2, label='observed ss mean')
axvline(f, color='g', linewidth=2, linestyle=':', label='input spike rate')
xlabel('synapse state', fontsize=18)
ylabel('P(syn)=synapse state', fontsize=18)
title('The input spike rate determines\nthe mean and variance of the synapse', fontsize=18);
```

```python
# validate neuron
dt = .001
N = 1000

t_low = 950.
t_high = 1050.
period = 5

neuron = ThresholdNeuron(dt , t_low, t_high, period)

inputs = np.linspace(900, 1100, 100)
spiked = np.zeros((N, len(inputs)))
for d_idx, d_in in enumerate(inputs):
    neuron.reset()
    for idx in xrange(N):
        if neuron.step(direct_in=d_in):
            spiked[idx, d_idx] = 1

spike_rates = np.sum(spiked, axis=0) / (float(N)*dt)

figure(figsize=(8,6))
plot(inputs, spike_rates, '-o')
xlabel('input current', fontsize=18)
ylabel('spike rate (Hz)', fontsize=18)
title('The threshold neuron outputs a constant spike rate in response\n' +
      'to a constant input within a specified range', fontsize=18);
```

```python
# simulate neuron and synapse together
dt = .001
n_steps = 10000
n_inputs = 30

tau = .01
period = 5

t_low = 600
t_high = 1400
w = 1.

neuron = ThresholdNeuron(dt, t_low, t_high, period, tau, w)
input_rates = np.linspace(0., t_high*2., n_inputs)
spike_rates = get_tuning(neuron, input_rates, n_steps)

th_spike_rates = np.zeros(n_inputs)
idx = np.logical_and(input_rates>=t_low, input_rates<=t_high)
th_spike_rates[idx]=1./(dt*period)

figure(figsize=(8,6))
plot(input_rates, spike_rates, '-o', label='observed spike rate')
plot(input_rates, th_spike_rates, 'r', label='constant input spike rate')
ylim(ylim()[0], 1.01/(dt*period))
legend(loc='upper right')
xlabel('input rate (Hz)', fontsize=18)
ylabel('spike rate (Hz)', fontsize=18)
title("The noisy synapse smooths out the neuron's tuning curve\n", fontsize=18);
```

```python
# simulate bump neuron
dt = .001
n_dim = 2
n_steps = 5
n_inputs_d = 50
n_inputs = n_inputs_d**n_dim

tau = .01
period = 5

t_low = 0.
t_high = .5

e_vector = np.ones(n_dim)
e_vector /= norm(e_vector)

# with direct current injection  #####################
direct_in_d = np.linspace(0., 1.8, n_inputs_d)
direct_inputs_x, direct_inputs_y = np.meshgrid(direct_in_d, direct_in_d)
direct_inputs = unroll_inputs(direct_inputs_x, direct_inputs_y)

neuron = BumpThresholdNeuron(e_vector, dt, t_low, t_high, period, tau)
spike_rates = get_tuning_multid(neuron, input_rates=None, direct_inputs=direct_inputs, n_steps=n_steps)
spike_rates = spike_rates.reshape((n_inputs_d, n_inputs_d))

fig = figure(figsize=(10,8))

ax = fig.add_subplot(111, projection='3d')
p = ax.plot_surface(direct_inputs_x, direct_inputs_y, spike_rates, rstride=1, cstride=1,
                    cmap = cm.coolwarm, linewidth=0, antialiased=False)
xlabel('x input', fontsize=18)
ylabel('y input', fontsize=18)
ax.set_zlabel('spike rate (Hz)', fontsize=18)
title('Multidimensional bump tuning using the\n' +
      'Threshold Neuron and direct current input', fontsize=18)

# with spikes #####################
n_steps = 2000
n_inputs_d = 30
n_inputs = n_inputs_d**n_dim

f_target = 1000.
w = 1./f_target
f_low = 0.
f_high = 1800.

in_rates_d = np.linspace(f_low, f_high, n_inputs_d)
in_rates_x, in_rates_y = np.meshgrid(in_rates_d, in_rates_d)
in_rates = unroll_inputs(in_rates_x, in_rates_y)

neuron = BumpThresholdNeuron(e_vector, dt, t_low, t_high, period, tau_syn=tau, w_syn=w)
spike_rates = get_tuning_multid(neuron, input_rates=in_rates, direct_inputs=None, n_steps=n_steps)
spike_rates = spike_rates.reshape((n_inputs_d, n_inputs_d))
fig = figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(in_rates_x, in_rates_y, spike_rates, rstride=1, cstride=1,
                    cmap = cm.coolwarm, linewidth=0, antialiased=True)
xlabel('x input (Hz)', fontsize=18)
ylabel('y input (Hz)', fontsize=18)
ax.set_zlabel('spike rate (Hz)', fontsize=18)
title('Multidimensional bump tuning using the\n' +
      'Threshold Neuron and spiking input', fontsize=18)
```
