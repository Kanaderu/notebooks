```python
class Neuron(object):
    def __init__(self, dt=100):
        self.tscale = 1e-6   # units of time (s)
        self.dt = dt         # dt = integration time step, in tscale
        self.updatetime = 0  # last update time

        self.u = 0           # u is the internal synaptic state
        self.w = .004           # w is the weight: scale factor
        self.syntau = 10000  # synapse tau, in tscale
        self.syn = 0.0       # syn is the synaptic input current to the soma
        
        self.x = 0.0         # neuron state
        self.neutau = 5000   # time scale of neuron integration in microseconds
        self.threshold = 10  # spiking threshold for neuron state x
        
        # override all the soma computation, just use constant inputs
        self.override = False
        
    def reset(self):
        self.u = 0
        self.updatetime = 0
        self.x = 0.
        self.syn = 0.
        
    def debug_syn(self,incur):
        self.override = True
        self.incur = incur

    def debug_stop(self):
        self.override = False
        
    # call if a spike arrives
    def inspike(self):
        self.u = self.u + 1
    
    # return 1 if there is a spike, 0 if there isn't.
    def step(self, curtime):
        if self.updatetime > curtime: return 0
        self.updatetime = self.updatetime + self.dt
        
        self.step_syn()
        self.step_soma()
       
        # spike detect
        if self.x > self.threshold:
            self.x = 0.0
            return 1
        else:
            return 0
    
    def step_syn(self):
        # synaptic update
        self.syn = self.syn - self.syn*float(self.dt)/self.syntau + self.w*float(self.u)/self.syntau/self.tscale
        self.u = 0
        
    def step_soma(self):
        # neuron update
        syn = self.incur if self.override else self.syn
        self.x = self.x + float(self.dt)/self.neutau*(0.5*(self.x**2) - self.x + syn)
                
    def readneuron(self):
        return self.x
    
    def readpostsyn(self):
        return self.syn


class DigitalNeuron(Neuron):
    def __init__(self, dt=100):
        super(DigitalNeuron, self).__init__(dt)
        
        # synapse digitization
        self.synibits = 4    # 4 bits for integer component
        self.synfbits = 12   # 12 bits for fractional component
        
        # neuron digitization
        self.neuibits = 4
        self.neufbits = 12
        
    def digitize (self,x,ibits,fbits):
        y = x*(1<<fbits)
        y = y % (1 << (fbits + ibits))
        return (int(y)+0.0)/(1<<fbits)
    
    def step_syn(self):
        super(DigitalNeuron, self).step_syn()
        self.syn = self.digitize(self.syn, self.synibits, self.synfbits)
        
    def step_soma(self):
        super(DigitalNeuron, self).step_soma()
        self.x = self.digitize(self.x, self.neuibits, self.neufbits)


class StochasticNeuron(DigitalNeuron):
    def __init__(self, dt=100, k=5):
        super(StochasticNeuron, self).__init__(dt)
        self.k = k  # number of increments to apply when spike arrives
        #self.pr = (self.dt+0.0)/(self.syntau)
        
        self.pr = np.exp(-float(self.dt)/self.syntau)
        self.synscale = (1.-self.pr)/(self.k*float(dt)*self.tscale)
        print self.pr, self.synscale
    
    def inspike(self):
        self.u = self.u + self.k
        
    def step_syn(self):
        self.u -= sum(np.random.rand(self.u) > self.pr)
        self.syn = self.w*self.u*self.synscale
        self.syn = self.digitize(self.syn, self.synibits, self.synfbits)
        
```

```python
class SpikeGen(object):
    def __init__(self, interval=5000):
        self.interval = interval # inter-spike interval (in microseconds)
        self.updatetime = 0  # next update time
        
    def reset(self):
        self.updatetime = 0
        
    def step(self, curtime):
        if curtime < self.updatetime: return 0
        self.updatetime = self.updatetime + self.interval
        return 1
```

```python
t=0
exp_dt = 1e-6
nsteps = int(1e5)
time = exp_dt * np.arange(nsteps)
n_analog = Neuron(dt=4)
n_digital = DigitalNeuron(dt=4)
n_stochastic = StochasticNeuron(dt=4, k=4)

n_analog.debug_syn(1.)
g = SpikeGen(2000)

inspike = []

# analog 
x_a = np.zeros(nsteps)
incurrent_a = np.zeros(nsteps)
outspike_a = []

# digital
x_d = np.zeros(nsteps)
outspike_d = []
incurrent_d = np.zeros(nsteps)

# stochastic
x_s = np.zeros(nsteps)
outspike_s = []
incurrent_s = np.zeros(nsteps)

for i in xrange(nsteps):
    x_a[i] = n_analog.readneuron()
    x_d[i] = n_digital.readneuron()
    x_s[i] = n_stochastic.readneuron()

    incurrent_a[i] = n_analog.readpostsyn()
    incurrent_d[i] = n_digital.readpostsyn()
    incurrent_s[i] = n_stochastic.readpostsyn()
    if g.step(t) == 1: 
        inspike.append(t*exp_dt)
        n_analog.inspike()
        n_digital.inspike()
        n_stochastic.inspike()
    if n_analog.step(t) == 1: outspike_a.append(t*exp_dt)
    if n_digital.step(t) == 1: outspike_d.append(t*exp_dt)
    if n_stochastic.step(t) == 1: outspike_s.append(t*exp_dt)
    t = t + 1
```

```python
figure('results', figsize=(8,6)).clf()

ax = subplot(4, 1, 1)
ax.set_color_cycle(['r','g','b'])
x0 = plot(time, x_a)
x1 = plot(time, x_d)
x2 = plot(time, x_s)
#ax.legend((x0[0], x1[0], x2[0]), ('Analog', 'Digital', 'Stochastic'), loc=2)
title('Neuron state')

ax = subplot(4,1,2)
ax.set_color_cycle(['r','g','b'])
plot(time, incurrent_a)
plot(time, incurrent_d)
plot(time, incurrent_s)
title('Synapse output')

px = subplot(4,1,3, sharex = ax)
px.set_color_cycle(['r','g','b'])
px.stem(inspike, [1]*len(inspike))
title ('Input spikes')

px = subplot(4,1,4, sharex=ax)
if len(outspike_a) != 0: px.stem(outspike_a, [1]*len(outspike_a), 'r', markerfmt='ro')
if len(outspike_d) != 0: px.stem(outspike_d, [0.75]*len(outspike_d), 'g', markerfmt='go')
if len(outspike_s) != 0: px.stem(outspike_s, [0.5]*len(outspike_s), 'b', markerfmt='bo')
title ('Output spikes')
xlabel('time(s)')

tight_layout()
show()
```

```python
# experiment: n = neuron + synapse model, inspk = spike generator, tm = duration in µsec
def experiment(n,inspk,tm):
    t=0
    r_total=range(tm)
    outcnt = 0
    for i in r_total:
        if inspk.step(tm) == 1:
            n.inspike()
        if n.step(tm) == 1: outcnt = outcnt + 1
    return (0.0 + outcnt)/(0.0+tm)

def many_exp(n,inspk,tm,count):
    x = 0
    x2 = 0
    sys.stdout.write('Experiment ' + str(count) + ': ')
    for i in range(count):
        sys.stdout.write('[' + str(i) + ']')
        sys.stdout.flush()
        n.reset()
        inspk.reset()
        v = experiment(n,inspk,tm)
        x = x + v
        x2 = x2 + v*v
    m = (x+0.0)/count
    sys.stdout.write("\n")
    return [m, math.sqrt(math.fabs((x2+0.0)/count-m**2))]
```
