# Notebook description

This notebook looks at the signal to noise ratio (SNR) of a synapse with Poisson input spike train with frequency $\lambda$.

$$SNR = \frac{E[X]}{\sqrt{\operatorname{Var}(X)}}$$

at the steady state where $X$ is a random variable describing the state of the synapse.

The synapse is a first order lowpass filter described by
$$\tau\dot{x}=-x+\sum_i\delta(t-t_i)$$

where $\tau$ is the synaptic time constant and $t_i$ describes the arrival time of the $i$th spike.

Our approach is to consider $X$ at an arbitrary point in time. The $i$th spike that arrived prior to this point in time contributes 

$$\frac{1}{\tau}e^{-t_i/\tau}$$

to the synapse's current state. Note that $t_i$ is now the time since the arrival of the $i$th spike. The value of the synapse is then

$$x=\sum_{i=1}^{\infty} \frac{1}{\tau}e^{-t_i/\tau}.$$

To deal with the $\infty$, we will first consider another random variable $Y$ that only looks back $k\tau$ back in time such that 

$$y=\sum_{i=1}^{N} \frac{1}{\tau}e^{-t_i/\tau}.$$

where $N$ is a random variable representing the number of spikes received within $k\tau$ of the present time.

From here 

$$
x = \lim_{k\rightarrow\infty}y
$$

# Summary of results

\begin{align}
E[X] &= \lambda \\
\operatorname{Var}(X) &= \frac{\lambda}{2\tau} \\
SNR &= \sqrt{2\lambda\tau} \\
\end{align}

# Derivations

## Synapse mean

\begin{aligned}
E[Y] &= E\left[\sum_{i=1}^{N}\frac{1}{\tau}e^{-t_i/\tau}\right] \\
 &= E\left[E\left[\sum_{i=1}^{N}\frac{1}{\tau}e^{-t_i/\tau}\bigg|N\right]\right]\qquad\text{law of total expectation} \\
 &= E\left[\sum_{i=1}^{N}E\left[\frac{1}{\tau}e^{-t_i/\tau}\bigg|N\right]\right] \\
 &= E\left[\sum_{i=1}^{N}E\left[\frac{1}{\tau}e^{-t_i/\tau}\right]\right] \\
 &= E\left[\sum_{i=1}^{N}1\right]E\left[\frac{1}{\tau}e^{-t_i/\tau}\right] \\
 &= E[N]E\left[\frac{1}{\tau}e^{-t_i/\tau}\right] \\
\end{aligned}

Since the impulse response is a decaying exponential, it never reaches 0. To avoid having to then consider all spikes since the beginning of time,  we make the simplifying assumption that the impulse response is negligible after a certain time. That is, we only consider spikes within $k\tau$ of the current time, where $k$ is an arbitrary, positive value. The higher $k$ is, the further back in time we consider. As $N$ is the number of spikes from a Poisson spike train with rate $\lambda$ received within $k\tau$, $E[N]=\lambda k\tau$. In addition, spike times within a time period are uniform for a Poisson spike train, so $t_i$ is uniform and

\begin{aligned}
E\left[\frac{1}{\tau}e^{-t_i/\tau}\right] &= \int_0^{k\tau}\frac{1}{k\tau}\frac{1}{\tau}e^{-t_i/\tau}dt_i \\
 &= \frac{1}{k\tau}\int_0^{k\tau}\frac{1}{\tau}e^{-t_i/\tau}dt_i \\
 &= \frac{1}{k\tau}\left(1-e^{-k\tau/\tau}\right) \\
 &=  \frac{1}{k\tau}\left(1-e^{-k}\right)
\end{aligned}

Therefore 

\begin{aligned}
E[Y] &= E[N] E\left[\frac{1}{\tau}e^{-t_i/\tau}\right] \\
 &= \lambda k\tau \frac{1}{k\tau}\left(1-e^{-k}\right) \\
 &= \lambda \left(1-e^{-k}\right)
\end{aligned}

Taking the limit as $k$ tends to infinity yields

\begin{aligned}
E[X] &= \lim_{k\rightarrow\infty}E[Y] \\
 &= \lim_{k\rightarrow\infty}\lambda \left(1-e^{-k}\right) \\
E[X] &=\lambda
\end{aligned}

as would be expected.

## Syanpse variance

By definition, 

$$\operatorname{Var}(X) = E[X^2] - E[X]^2$$

We've determined $E[X]$ already, so let's find $E[X^2]$.

\begin{aligned}
E[Y^2] &= E\left[\left(\sum_{i=1}^{N}\frac{1}{\tau}e^{-t_i/\tau}\right)^2\right] \\
 &= E\left[\left(\sum_{i=1}^{N}\frac{1}{\tau}e^{-t_i/\tau}\right)\left(\sum_{j=1}^{N}\frac{1}{\tau}e^{-t_j/\tau}\right)\right] \\
 &= E\left[\frac{1}{\tau^2}\left(e^{-t_0/\tau}+e^{-t_1/\tau}+\ldots+e^{-t_N/\tau}\right)
  \left(e^{-t_0/\tau}+e^{-t_1/\tau}+\ldots+e^{-t_N/\tau}\right)\right] \\
 &= E\left[\frac{1}{\tau^2}\left(\overbrace{\sum_{i=0}^{N-1}e^{-2t_i/\tau}}^{\text{squared terms}} + 
  \overbrace{\sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}e^{-t_i/\tau}e^{-t_j/\tau}}^{\text{cross terms}}\right)\right] \\
 &= E\left[\frac{1}{\tau}\sum_{i=0}^{N-1}\frac{1}{\tau}e^{-2t_i/\tau}+
  \sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\right] \\
 &= E\left[\frac{1}{\tau}\sum_{i=0}^{N-1}\frac{1}{\tau}e^{-2t_i/\tau}\right]+
  E\left[\sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\right] \\
\end{aligned}

First, let's consider the squared terms. This is very similar to the derivation for $E[Y]$

\begin{aligned}
E\left[\frac{1}{\tau}\sum_{i=0}^{N-1}\frac{1}{\tau}e^{-2t_i/\tau}\right] &= 
 \frac{1}{\tau}E\left[\sum_{i=0}^{N-1}\frac{1}{\tau}e^{-2t_i/\tau}\right] \\
 &= \frac{1}{\tau}E\left[E\left[\sum_{i=0}^{N-1}\frac{1}{\tau}e^{-2t_i/\tau}\bigg|N\right]\right] \\
 &= \frac{1}{\tau}E[N]E\left[\frac{1}{\tau}e^{-2t_i/\tau}\right] \\
 &= \frac{1}{\tau}(\lambda k\tau)\int_0^{k\tau}\frac{1}{k\tau}\frac{1}{\tau}e^{-2t_i/\tau}dt_i \\
 &= \frac{\lambda}{\tau}\int_0^{k\tau}\frac{1}{\tau}e^{-2t_i/\tau}dt_i \\
 &= \frac{\lambda}{2\tau}\left(1-e^{-2k\tau/\tau}\right) \\
 &= \frac{\lambda}{2\tau}\left(1-e^{-2k}\right) \\
\end{aligned}

Now let's consider the consider the cross terms.

\begin{aligned}
E\left[\sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\right] &=
   E\left[E\left[\sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\bigg|N\right]\right] \\
  &= E\left[\sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}E\left[\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\bigg|N\right]\right] \\
  &= E\left[\sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}E\left[\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\right]\right] \\
  &= E\left[\sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}1\right]E\left[\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\right] \\
  &= E\left[N^2-N\right]E\left[\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\right] \\
  &= \left(E\left[N^2\right]-E[N]\right)E\left[\frac{1}{\tau}e^{-t_i/\tau}\right]E\left[\frac{1}{\tau}e^{-t_j/\tau}\right]
   \quad i\neq j \ \text{ so } t_i, t_j \text{ are independent} \\
  &= \left(E\left[N^2\right]-E[N]\right)\frac{1}{k\tau}\left(1-e^{-k}\right)\frac{1}{k\tau}\left(1-e^{-k}\right) \\
  &= \left(E\left[N^2\right]-E[N]\right)\frac{1}{k^2\tau^2}\left(1-e^{-k}\right)^2 \\
\end{aligned}

For poisson random variable $N$ with parameter $\lambda k\tau$, $\operatorname{Var}(N)=E[N]=\lambda k\tau$

\begin{aligned}
\lambda k\tau &= E\left[N^2\right]-E[N]^2 \\
 &= E\left[N^2\right] - (\lambda k\tau)^2 \\
E\left[N^2\right] &= \lambda k\tau + (\lambda k\tau)^2 
\end{aligned}

Returning to the Expected cross term derivation

\begin{aligned}
E\left[\sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\right] &=
    \left(E\left[N^2\right]-E[N]\right)\frac{1}{k^2\tau^2}\left(1-e^{-k}\right)^2 \\
 &= \left(\lambda k\tau + (\lambda k\tau)^2 - \lambda k\tau\right)\frac{1}{k^2\tau^2}\left(1-e^{-k}\right)^2 \\
 &= \left(\lambda k\tau\right)^2\frac{1}{k^2\tau^2}\left(1-e^{-k}\right)^2 \\
 &= \lambda^2\left(1-e^{-k}\right)^2 \\
\end{aligned}

Putting the Expectation of the squared terms with the Expectation of the cross terms,

\begin{aligned}
E\left[Y^2\right] &= E\left[\frac{1}{\tau}\sum_{i=0}^{N-1}\frac{1}{\tau}e^{-2t_i/\tau}\right] +
    E\left[\sum_{i=0,i\neq j}^{N-1}\sum_{j=0}^{N-1}\frac{1}{\tau}e^{-t_i/\tau}\frac{1}{\tau}e^{-t_j/\tau}\right] \\
 &= \frac{\lambda}{2\tau}\left(1-e^{-2k}\right) + \lambda^2\left(1-e^{-k}\right)^2,
\end{aligned}

and relating to $E[X^2]$,

\begin{aligned}
E[X^2] &= \lim_{k\rightarrow\infty}E[Y^2] \\
 &= \lim_{k\rightarrow\infty} \frac{\lambda}{2\tau}\left(1-e^{-2k}\right) + \lambda^2\left(1-e^{-k}\right)^2 \\
E[X^2] &= \frac{\lambda}{2\tau} + \lambda^2
\end{aligned}

Plugging into the variance equation,

\begin{aligned}
\operatorname{Var}(X) &= E[X^2] - E[X]^2 \\
 &= \frac{\lambda}{2\tau} + \lambda^2 - \lambda^2 \\
\operatorname{Var}(X) &= \frac{\lambda}{2\tau}
\end{aligned}

## Synapse SNR

\begin{aligned}
SNR &= \frac{\mu}{\sigma} \\
 &= \frac{\lambda}{\sqrt{\lambda/2\tau}} \\
 &= \sqrt{2\lambda\tau}
\end{aligned}

```python
# initialize pylab and constants
%pylab inline
import numpy as np

s = random.seed(2)
rng = random.RandomState(s)
```

```python
#theory functions
def th_mean(lam):
    return lam

def th_mean_sq(lam, tau):
    return lam/(2.*tau) + lam**2

def th_var(lam, tau):
    return lam/(2.*tau)

def th_snr(lam, tau):
    return np.sqrt(2*lam*tau)

# simulation functions
def spike_value(t, tau):
    return 1./tau*exp(-t/tau)

def get_snr(mu, var):
    return mu/np.sqrt(var)
```

```python
tau = .01
lam = np.linspace(1,1000,100)
e_x = th_mean(lam)
var_x = th_var(lam, tau)
snr_x = th_snr(lam, tau)

fig = figure(figsize=(8,12))
subplot(311)
plot(lam, e_x, 'b')
ylabel(r'$E[x]$', fontsize=20)
title(r'$\tau=%.1f$ ms' % (tau*1000.), fontsize=20)
subplot(312)
plot(lam, var_x, 'b')
ylabel(r'$\operatorname{Var}(x)$', fontsize=20)
subplot(313)
plot(lam, snr_x, 'b')
ylabel(r'$SNR$', fontsize=20)
xlabel(r'$\lambda$ (spikes per second)', fontsize=20);
```

# Fixed Rate Simulation

```python
def sim_distribution(k=30., tau=.01, lam=1000., trials=4000):
    # k number of time constants to consider
    # tau time constant of the filter
    # lam frequency of spikes in
    nspikes = k*tau*lam

    isi = rng.exponential(1./lam, (trials, nspikes))
    spike_times = np.cumsum(isi, axis=1)
    spike_vals = spike_value(spike_times, tau)
    Xs = np.sum(spike_vals, axis=1)

    fig=figure(figsize=(12,6))
    ax=fig.add_subplot(111)
    ret=ax.hist(Xs/lam, bins=max(10, trials/100), normed=True, histtype='step')

    mean_obs = mean(Xs)
    mean_th = th_mean(lam)
    axvline(mean_obs/lam, color='r', label='observed mean')
    axvline(mean_th/lam, color='k', linestyle='--', label='theoretical mean')

    var_obs = var(Xs)
    var_th = th_var(lam, tau)
    std_th = np.sqrt(var_th)

    for n in [1.,2.]:
        line = axvline((mean_th+n*std_th)/lam, color='g', linestyle=':', linewidth=2)
        line = axvline((mean_th-n*std_th)/lam, color='g', linestyle=':', linewidth=2)
    a, astep = np.linspace(xlim()[0], xlim()[1], 100, retstep=True)
    def normal_pdf(x, mu=0., sigma=1.):
        return 1./(sigma*np.sqrt(2*pi))*np.exp(-(x-mu)**2/(2*sigma**2))
    normal_approx = normal_pdf(a, mean_th/lam, std_th/lam)
    plot(a, normal_approx, 'g', label='Gaussian approx')
    line.set_label(r'Gaussian $\mu \pm n\sigma$')

    from scipy.special import erf
    def normal_cdf(x, mu=0., sigma=1.):
        return 1./2.*(1+erf((x-mu)/(sigma*np.sqrt(2))))
    for n in [1.,2.]:
        percentiles = [normal_cdf(-n)*100., normal_cdf(n)*100.]
        pct_obs = np.percentile(Xs, percentiles)
        line = axvline(pct_obs[0]/lam, color='r', linestyle=':')
        line = axvline(pct_obs[1]/lam, color='r', linestyle=':')
    line.set_label(r'observed $\sigma$ percentiles')

    ax.set_title(r'$\lambda=%.0f$ Hz, $\tau=%.1f$ ms, $%d$ trials' % (lam, tau*1000., trials), fontsize=20)
    ax.set_ylabel('normalized counts', fontsize=16)
    ax.set_xlabel(r'$E[X]/\lambda$', fontsize=20)
    ax.legend(loc='upper right');

sim_distribution(k=30., tau=.01, lam=10., trials=40000)
sim_distribution(k=30., tau=.01, lam=100., trials=40000)
sim_distribution(k=30., tau=.01, lam=1000., trials=40000)
sim_distribution(k=30., tau=.01, lam=10000., trials=40000)
```

# Variable Rate Simulations

```python
def check_theory(lams, k=30., tau=.01, trials=300):
    # lams array of frequencies to test
    # k number of time constants to consider in each experiment
    # tau time constant of the filter
    # lam frequency of spikes innexp = 300 # number of experiments
    nspikes = 1000 # number of spikes in each experiment
    nfreqs = len(lams) # number of frequencies to test

    obs_mean = np.zeros(nfreqs)
    obs_var = np.zeros(nfreqs)
    obs_snr = np.zeros(nfreqs)
    for lam_idx, lam in enumerate(lams):
        nspikes = k*tau*lam
        isi = rng.exponential(1./lam, (trials, nspikes))
        spike_times = np.cumsum(isi, axis=1)
        spike_vals = spike_value(spike_times, tau)
        exp_state = np.sum(spike_vals, axis=1)
        obs_mean[lam_idx] = np.mean(exp_state)
        obs_var[lam_idx] = np.var(exp_state)
        obs_snr[lam_idx] = get_snr(obs_mean[lam_idx], obs_var[lam_idx])

    mean_th = th_mean(lams)
    var_th = th_var(lams, tau)
    snr_th = th_snr(lams, tau)

    figure(figsize=(12,9))
    subplot(311)
    plot(lams, obs_mean, 'b', label='simulation')
    plot(lams, mean_th, 'r--', label='theory')
    title(r'$\tau=%.1f$ ms' % (tau*1000.), fontsize=20)
    ylabel(r'$E[x]$', fontsize=20)
    legend(loc='upper left')
    subplot(312)
    plot(lams, obs_var, 'b')
    plot(lams, var_th, 'r--')
    ylabel(r'$\operatorname{Var}(x)$', fontsize=20)
    subplot(313)
    plot(lams, obs_snr, 'b')
    plot(lams, snr_th, 'r--')
    ylabel(r'$SNR(x)$', fontsize=20)
    xlabel(r'$\lambda$ (spikes per second)', fontsize=20)
    
lams = np.linspace(5, 1000, 50)
check_theory(lams, k=20, trials=10000)
```
