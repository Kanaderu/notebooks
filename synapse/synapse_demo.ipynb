# Notebook description

This notebook demonstrates the exponential behaviour of the synapse and its noisiness when receiving spiking input. Demoed is the state of a synapse with fixed time constant as it receives Poisson and uniform spikes at varying rates.

The synapse is a first order lowpass filter described by

$$\tau\dot{x}=-x+\sum_i\delta(t-t_i)$$

where $\tau$ is the synaptic time constant and $t_i$ describes the arrival time of the $i$th spike.

The input spike rate is $\lambda$, and the spike times are drawn from either a Poisson or uniform process.

# Demo

```python
# initialize pylab and constants
%pylab inline
import numpy as np
```

```python
# demo synapse
pstc = .05   # post synaptic time constant
T = 10*pstc      # total simulation time

figure("state_timeseries", figsize=(16,13))

rates = np.logspace(-1,3,9)/pstc
for idx, f in enumerate(rates):
    dt = 1./(f*100.)
    Tf_uniform = 1./f   # firing rate period
    Tf_poisson = np.random.exponential(1./f) # interspike interval period
    elapsed_uniform = Tf_uniform # elapsed time since last spike
    elapsed_poisson = Tf_poisson # elapsed time since last spike
    
    steps = int(T/dt)
    decay = np.exp(-dt/pstc)
    
    t = 0.
    time = np.arange(steps)*dt
    time_norm = time / pstc
    state_poisson = np.zeros(steps)
    state_uniform = np.zeros(steps)
    
    for i in range(steps-1):
        state_poisson[i+1] = state_poisson[i]*decay
        if elapsed_poisson >= Tf_poisson:
            state_poisson[i+1] += (1-decay)/dt
            elapsed_poisson = elapsed_poisson - Tf_poisson + dt
            Tf_poisson = np.random.exponential(1./f)
        else:
            elapsed_poisson += dt

        state_uniform[i+1] = state_uniform[i]*decay
        if elapsed_uniform >= Tf_uniform:
            state_uniform[i+1] += (1-decay)/dt
            elapsed_uniform = elapsed_uniform - Tf_uniform + dt
        else:
            elapsed_uniform += dt
    
    x_u = 1./(pstc*(1.-np.exp(-1./(f*pstc))))
    x_l = np.exp(-1./(f*pstc))*x_u
    
    # plot timeseries
    subplot(3,3,idx+1)
    plot(time_norm, state_poisson/f, 'r', label='Poisson input')
    plot(time_norm, state_uniform/f, 'b', label='Uniform input')
    axhline(x_u/f, c='b', ls='--', label='SS Uniform max/min')
    axhline(x_l/f, c='b', ls='--')
    for i in xrange(6):
        axvline(i, c='k', ls=':')
        axhline((1-np.exp(-i)), c='k', ls=':')
    title(r'$\lambda\tau=%3.1f$' % (f*pstc), fontsize=20)

subplot(331)
legend(loc='upper right')
subplot(338)
xlabel(r'Time ($\tau$)', fontsize=18)
subplot(334)
ylabel('$x/\lambda$', fontsize=20)
tight_layout()
```

The plots are normalized in time to units of $\tau$ and are normalized in synaptic value by $\lambda$. The input spike rates are expressed as $\lambda\tau$, the expected number of spikes that are input to the synapse in a single time constant.

Here are a few points to observe from the plots:
 - The synapse reaches a steady state value equal to the input firing rate in expectation
 - The speed at which the synapse value approaches the input firing rate value is determined by the synaptic time constant
 - The synapse with uniform spiking input is much less noisy than the synapse with Poisson spiking input. The synapse with uniform input is, in fact, bounded in steady state. These points made more precisely when comparing

 on Poisson input spiking and

 on uniform input spiking
